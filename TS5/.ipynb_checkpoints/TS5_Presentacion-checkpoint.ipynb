#### <span style="text-decoration: underline;">Obtención de frecuencias de corte inferior y superior:</span>

#####
##### Referimos a la ecuación 9.16 del libro de Schaumann, Van Valkenburg, *"Design of Analog Filters"*, en adelante mencionado sencillamente como "el libro de Schaumann" [$^{(1)}$](#1):
#####

\begin{equation}
\omega_0= \sqrt{\omega_1*\omega_2}
\end{equation}

#####
##### Reescribimos la ecuación de $\omega_2$ en función de $\omega_0$ y $\omega_1$:
#####

\begin{equation}
\omega_2= \frac{\omega_0^2}{\omega_1}
\end{equation}

#####
##### Referimos a la ecuación 9.19 del libro de Schaumann [$^{(2)}$](#2):
#####

\begin{equation}
Q= \frac{\omega_0}{B}= \frac{1}{\frac{\omega_2}{\omega_0}-\frac{\omega_1}{\omega_0}} 
\end{equation}

#####
##### Reescribimos la ecuación de $\omega_2$ en función de $\omega_0$ y $\omega_1$:
#####

\begin{equation}
Q= \frac{1}{\frac{\omega_2-\omega_1}{\omega_0}} 
\end{equation}

#####

\begin{equation}
Q= \frac{\omega_0}{\omega_2-\omega_1} 
\end{equation}

#####

\begin{equation}
\omega_2-\omega_1= \frac{\omega_0}{Q} 
\end{equation}

#####

\begin{equation}
\omega_2= \frac{\omega_0}{Q}+\omega_1
\end{equation}

#####
##### Resolvemos el sistema de ecuaciones que resulta de la igualdad de ambas ecuaciones de $\omega_2$:
#####

\begin{equation}
\frac{\omega_0^2}{\omega_1}= \frac{\omega_0}{Q}+\omega_1
\end{equation}

#####

\begin{equation}
\frac{\omega_0^2}{\omega_1}-\omega_1= \frac{\omega_0}{Q}
\end{equation}

#####
##### Multiplicamos a ambos miembros por $\omega_1$:
#####

\begin{equation}
\left(\frac{\omega_0^2}{\omega_1}-\omega_1\right)*\omega_1= \frac{\omega_0}{Q}*\omega_1
\end{equation}

#####

\begin{equation}
\omega_0^2 - \omega_1^2= \frac{\omega_0}{Q}*\omega_1
\end{equation}

#####

\begin{equation}
\omega_1^2+ \frac{\omega_0}{Q}*\omega_1 -\omega_0^2= 0
\end{equation}

#####
##### Reemplazamos valores y resolvemos la ecuación cuadrática:
#####

\begin{equation}
\omega_1^2+ \frac{2*\pi*22kHz}{5}*\omega_1 -(2*\pi*22kHz)^2= 0
\end{equation}

#####

\begin{equation}
\omega_1^2+ 8800*\pi*\left[\frac{rad}{s}\right]*\omega_1- 1,936*10^9*\pi^2\left[\frac{rad}{s}\right]^2= 0
\end{equation}

#####

\begin{equation}
\omega_{11}= 125,0965*10^3 \left[\frac{rad}{s}\right]
\end{equation}

#####

\begin{equation}
\omega_{12}= \text{- ___________} \left[\frac{rad}{s}\right] \rightarrow descartado
\end{equation}

#####
##### Reemplazamos $\omega_{11}=\omega_1$ en una de las ecuaciones de $\omega_2$ y obtenemos en consecuencia:
#####

\begin{equation}
\omega_2= \frac{\omega_0^2}{\omega_1}= \frac{(2*\pi*22kHz)^2}{125,0965*10^3\left[\frac{rad}{s}\right]} 
\end{equation}

#####

\begin{equation}
\omega_2= 152,7425*10^3 \left[\frac{rad}{s}\right] 
\end{equation}

\begin{equation}
\Omega= Q*\frac{\omega^2-1}{\omega}

#### <span style="text-decoration: underline;">Cálculo de $\epsilon$, grado de libertad de la función, y $N$, el orden del filtro pasa bajos</span>

#####
##### Tomando como referencia las ecuaciones de $|T(j\Omega)|^2$ en su aproximación de Chevyshev[$^{(7)}$](#7)[$^{(8)}$](#8) según requerimiento del enunciado:
#####

\begin{equation}
|T(j\Omega)|^2= T(j\Omega)*T(-j\Omega)= {T(s)*T(-s)}\Bigr|_{s= j\Omega}= \frac{1}{1+\epsilon^2*C_n^2(\Omega)}
\end{equation}

#####
##### Podemos entonces establecer la atenuación en función de la ecuación anterior como la inversa de $|T(j\Omega)|^2$:
#####

\begin{equation}
|\alpha|^2= \frac{1}{|T(j\Omega)|^2}= \frac{1}{\frac{1}{1+\epsilon^2*C_n^2(\Omega)}}= 1+\epsilon^2*C_n^2(\Omega)
\end{equation}

#####
##### Considerando que $|\alpha|_{dB} \triangleq \alpha_{dB}$, podemos expresar la ecuación en dB [$^{(9)}$](#9):
#####

\begin{equation}
\alpha_{dB}= 10*log\left[1+\epsilon^2*C_n^2(\Omega)\right]
\end{equation}

#####
##### Expresamos $\epsilon^2$ como resultado de la ecuación anterior [$^{(10)}$](#10):
#####

\begin{equation}
\epsilon^2= \frac{10^{\frac{\alpha_{máx}}{10}}-1}{C_n^2(\Omega)}
\end{equation}

#####
##### Considerando la máxima atenuación requerida por la plantilla $\alpha_{máx}$ en el vértice de fin de la banda de paso $\Omega_p$, obtenemos $\epsilon^2$:
#####

\begin{equation}
\epsilon^2= \frac{10^{\frac{\alpha_{máx}}{10}}-1}{C_n^2(\Omega_p)}= \frac{10^{\frac{20}{10}}-1}{1}
\end{equation}

#####

\begin{equation}
\boxed{ε^2= 0,1220}
\end{equation}

#####

\begin{equation}
\boxed{ε= 0,3493}
\end{equation}

#####
##### Nuevamente debido a que $\alpha_{min}$ es distinto para cada una de las bandas de detenida, debemos obtener $N$ para cada uno de los casos:
#####
##### Considerando la mínima atenuación requerida por la plantilla en el vértice del inicio de la banda de detenida correspondiente al par ordenado $\left(\alpha_{mín_{1}}, \Omega_{s_{1}}\right)$, iteramos N de forma tal que la atenuación sea mayor o igual a $\alpha_{mín_{1}}$:
#####

\begin{equation}
\alpha_{mín} = 10*log\left[1 + \epsilon^2*C_n^2(\Omega_{s_{1}})\right]
\end{equation}

#####

\begin{equation}
N= 1: ¿\alpha(\Omega_{s_{1}}) = 2,6230 dB \geq α_{mín}= 16 dB? \rightarrow No...
\end{equation}

#####

\begin{equation}

\end{equation}

#####

\begin{equation}

\end{equation}

#####
##### Realizamos lo propio para el par ordenado $\left(\alpha_{mín_{2}}, \Omega_{s_{2}}\right)$:
#####

\begin{equation}
N= 1: ¿\alpha(\Omega_{s_{2}}) = 6,2390 dB \geq α_{mín}= 24 dB? \rightarrow No...

\begin{equation}

##### Finalmente se elige aquel $N$ que sea el mayor de los dos obtenidos de forma tal que satisfaga ambos requerimientos simultáneamente[$^{(11)}$](#11):

\begin{equation}

#### <span style="text-decoration: underline;">Cálculo de la función transferencia $T_{LP}(s)$</span>
#####

##### Recordando que $\left|T(j\Omega)\right|^2= T(j\Omega)*T(-j\Omega)= {T(s)*T(-s)}\Bigr|_{s= j\Omega}= \frac{1}{1+\epsilon^2*C_n^2(\Omega)}$ y habiendo obtenido $\epsilon^2$ y $N$, procedemos a calcular los polinomios de orden $N$ de Chevyshev definidos como:
#####

\begin{equation}
C_n^2= cosh^2\left[n*cosh^{-1}(\Omega_{s})\right]
\end{equation}

#####
##### Utilizando identidades trigonómetricas [$^{(12)}$](#12), calculamos el polinomio para el grado $N$ obtenido:
#####

\begin{equation}
C_0(\Omega)= 1 
\end{equation}

#####

\begin{equation}
C_1(\Omega)= \Omega
\end{equation}

#####

\begin{equation}
C_n(\Omega)= 2*\Omega*C_{n-1}(\Omega) - C_{n-2}(\Omega) 
\end{equation}

#####

\begin{equation}
C_2(\Omega)= 2*\Omega*\Omega - 1= 2*\Omega^2-1 
\end{equation}

#####

\begin{equation}
C_3(\Omega)= 2*\Omega\left(2*\Omega^2-1\right) - \Omega= 4*\Omega^3 -2*\Omega -\Omega= 4*\Omega^3 -3*\Omega  
\end{equation}

#####
##### Reemplazando el polinomio de grado 3 en la ecuación de $|T(j\Omega)|^2= \frac{1}{1+\epsilon^2*C_n^2(\Omega)}$, operamos algebraicamente:
#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{1}{1+\epsilon^2*C_3^2(\Omega)}
\end{equation}

#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{1}{1+\epsilon^2*(4*\Omega^3 -3*\Omega)^2}
\end{equation}

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{1}{1+\epsilon^2*(4*\Omega^3 -3*\Omega)*(4*\Omega^3 -3*\Omega)}
\end{equation}

#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{1}{1+\epsilon^2*(16*\Omega^6 -12*\Omega^4 -12*\Omega^4 +9*\Omega^2)}
\end{equation}

#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{1}{16*\epsilon^2*\Omega^6 -24*\epsilon^2*\Omega^4 +9*\epsilon^2*\Omega^2 +1}
\end{equation}

#####
##### Dejamos mónico el polinomio denominador dividiendo numerador y denominador por $16*\epsilon^2$:
#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{\frac{1}{16*\epsilon^2}} {\frac{16*\epsilon^2}{16*\epsilon^2}*\Omega^6 -\frac{24*\epsilon^2}{16*\epsilon^2}*\Omega^4 +\frac{9*\epsilon^2}{16*\epsilon^2}*\Omega^2 +\frac{1}{16*\epsilon^2}}
\end{equation}

#####
##### Definimos los coeficientes $a$, $b$ y $c$ como:
#####

\begin{equation}
a= \frac{1}{16*\epsilon^2}
\end{equation}

#####

\begin{equation}
b= \frac{24*\epsilon^2}{16*\epsilon^2}= \frac{3}{2}
\end{equation}

#####

\begin{equation}
c= \frac{9*\epsilon^2}{16*\epsilon^2}= \frac{9}{16}
\end{equation}

#####
##### Obtenemos $\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}$ en función de los coeficientes $a$, $b$ y $c$:
#####

\begin{equation}
\left|T(j\Omega)\right|^2_{\Omega=\frac{$}{j}}= \frac{a} {\Omega^6 -b*\Omega^4 +c*\Omega^2 +a}
\end{equation}

#####
##### Reemplazando $\Omega$ por $\frac{$}{j}$, obtenemos:
#####

\begin{equation}
T($)*T(-$)= \frac{a} {\left(\frac{$}{j}\right)^6 -b*\left(\frac{$}{j}\right)^4 +c*\left(\frac{$}{j}\right)^2 +a}
\end{equation}
 
\begin{equation}
T($)*T(-$)= \frac{a} {-$^6 -b*$^4 -c*$^2 +a}
\end{equation}

#####
##### Expresamos la ecuación de forma general:
#####

\begin{equation}
T($)*T(-$)= \frac{\alpha}{$^3 + \beta*$^2 + \gamma*$ + \alpha}*\frac{\alpha}{-$^3 + \beta*$^2 + -\gamma*$ + \alpha}
\end{equation}

#####

\begin{equation}
T($)*T(-$)= \frac{\alpha^2}{-$^6 +\beta*$^5 -\gamma*$^4 + \alpha*$^3 -\beta*$^5 +\beta^2*$^4 -\beta*\gamma*$^3 +\alpha*\beta*$^2 -\gamma*$^4 +\beta*\gamma*$^3 -\gamma^2*$^2 +\alpha*\gamma*$ -\alpha*$^3 +\alpha*\beta*$^2 -\alpha*\gamma*$ +\alpha^2}
\end{equation}

#####

\begin{equation}
T($)*T(-$)= \frac{\alpha^2}{-$^6 +(\beta- \beta)*$^5 +(-\gamma +\beta^2 -\gamma)*$^4 +(\alpha -\beta*\gamma +\beta*\gamma -\alpha)*$^3 +(\alpha*\beta -\gamma^2 +\alpha*\beta)*$^2 +(\alpha*\gamma -\alpha*\gamma)*$ +\alpha^2}
\end{equation}

#####

\begin{equation}
T($)*T(-$)= \frac{\alpha^2}{-$^6 +(\beta^2 -2*\gamma)*$^4 +(2*\alpha*\beta -\gamma^2)*$^2 +\alpha^2}
\end{equation}

#####
##### Igualamos los coeficientes de ambas expresiones y resolvemos el sistema de ecuaciones:
#####

\begin{equation}
a= \alpha^2= \frac{1}{16*\epsilon^2}
\end{equation}

#####

\begin{equation}
\alpha= \frac{1}{4*\epsilon}= \frac{1}{4*0,3493}= 0,7157 
\end{equation}

#####

\begin{equation}
-b= \beta^2 -2*\gamma= -\frac{3}{2}
\end{equation}

#####

\begin{equation}

\end{equation}

#####

\begin{equation}
-c= 2*\alpha*\beta -\gamma^2= -\frac{16}{9}
\end{equation}

#####

\begin{equation}
-c= 2*\alpha*\beta -\gamma^2= -\frac{16}{9}
\end{equation}

#####

\begin{equation}

\end{equation}

#####
##### Reemplazamos el valor de $\alpha$ en la ecuación de $\beta$:

\begin{equation}

\end{equation}

#####
##### Reemplazamos la ecuación de $\gamma$ en la de $\beta$:

#####

\begin{equation}

\end{equation}

#####

\begin{equation}
\beta= 0,6986*\left( \frac{\beta^4}{4} +\frac{3}{4}*\beta^2 +\frac{9}{16}\right) -0,3930
\end{equation}

#####

\begin{equation}
\beta= 0,6986*\frac{\beta^4}{4} +0,6986*\frac{3}{4}*\beta^2 +0,6986*\frac{9}{16} -0,3930 
\end{equation}

#####

\begin{equation}
0,1747*\beta^4 +0,5239*\beta^2 -\beta +0,3930 -0,3930= 0

#####

\begin{equation}

#####

#####

\begin{equation}
\beta_1= 0 \; , \quad \beta_2= 1,2529 \; , \quad \beta_3= u +jv \; , \quad \beta_4= u -jv  
\end{equation}

#####
##### Reemplazando $\beta_2= \beta= 1,2529$ en la ecuación de $\gamma$:
#####

\begin{equation}
\gamma= \frac{\beta^2}{2} +\frac{3}{4}= \frac{1,2529^2}{2} +\frac{3}{4}= 1,5349
\end{equation}

#####
##### Podemos expresar entonces la transferencia pasa bajos prototipo normalizada $T_{LP}(\$)$:
#####

\begin{equation}
\boxed{T_{LP}($)= \frac{\alpha}{$^3 + \beta*$^2 + \gamma*$ + \alpha}= \frac{0,7157}{$^3 + 1,2529*$^2 + 1,5349*$ + 0,7157}}
\end{equation}

#####
##### Aplicamos el núcleo de la transformación $\$= K(s)= Q*\frac{s^2+1}{s}$ para obtener la transferencia pasa banda objetivo normalizada $T_{BP}(s)$:
#####

\begin{equation}
T_{LP}(\$)\Bigr|_{\$= K(s)= Q*\frac{s^2+1}{s} }= T_{BP}(s)
\end{equation}

#####

\begin{equation}
T_{BP}(s)= \frac{0,7157}{\left(Q*\frac{s^2+1}{s}\right)^3 + 1,2529*\left(Q*\frac{s^2+1}{s}\right)^2 + 1,5349*\left(Q*\frac{s^2+1}{s}\right) + 0,7157}
\end{equation}

#####

\begin{equation}
T_{BP}(s)= \frac{0,7157}{Q^3*\frac{(s^2+1)^3}{s^3} + 1,2529*Q^2*\frac{(s^2+1)^2}{s^2} + 1,5349*\left(Q*\frac{s^2+1}{s}\right) + 0,7157}
\end{equation}

#####

\begin{equation}
T_{BP}(s)= \frac{0,7157}{Q^3*\frac{s^6+3*s^4+3*s^2+1}{s^3} + 1,2529*Q^2*\frac{s^4+2*s^2+1}{s^2} + 1,5349*\left(Q*\frac{s^2+1}{s}\right) + 0,7157}
\end{equation}

#####

\begin{equation}
T_{BP}(s)= \frac{0,7157}{\frac{Q^3*(s^6+3*s^4+3*s^2+1) +1,2529*Q^2*(s^4+2*s^2+1)*s +1,5349*Q*(s^2+1)*s^2 +0,7157*s^3}{s^3} + 1,2529*Q^2*\frac{s^4+2*s^2+1}{s^2} + 1,5349*\left(Q*\frac{s^2+1}{s}\right) + 0,7157}
\end{equation}

#####



\begin{equation}
T(s)= \frac {\frac{1}{2*ε}}{(\frac{1}{s})^2 + \sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*(\frac{1}{s}) + \frac{\sqrt{ε^2+1}}{2*ε}} 
\end{equation}

\begin{equation}
T(s)= \frac{\frac{1}{2*ε}}{\frac{1}{s^2} + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{s} + \frac{\sqrt{ε^2+1}}{2*ε}}
\end{equation}

\begin{equation}
T(s)= \frac{\frac{1}{2*ε}}{\frac{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}{s^2}}= \frac{\frac{1}{2*ε}*s^2}{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}
\end{equation}

\begin{equation}
T(s)= \frac{\frac{\frac{1}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2}{\frac{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}{\frac{\sqrt{ε^2+1}}{2*ε}}}
\end{equation}

\begin{equation}
T(s)= \frac {\frac{\frac{1}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2} {\frac{1}{\frac{\sqrt{ε^2+1}}{2*ε}} +\frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s +\frac{\frac{\sqrt{ε^2+1}}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2}
\end{equation}

\begin{equation}
T(s)= \frac {\frac{1}{\sqrt{ε^2+1}}*s^2} {\frac{2*ε}{\sqrt{ε^2+1}} +\frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s +s^2}
\end{equation}

\begin{equation}
T_{HP}(s)=\frac{\frac 1{\sqrt{ε^2+1}}*s^2}{s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

\begin{equation}
\boxed{T_{HP}(s)= \frac {0,8913*s^2} {s^2 +0,9957*s +0,9069}}
\end{equation}

#####
##### Considerando $T_{HP}(s)$ en términos de $\omega_0$ y $Q$, podemos expresarla de forma general de la siguiente manera:
#####

\begin{equation}
T_{HP}(s)=\frac{k * s^2} {s^2 + \frac{ω_0}{Q}*s + ω_0^2}
\end{equation}


#####
##### Obtenemos los valores de $\omega_0$ y $Q$ para este caso particular:
#####

\begin{equation}
ω_0^2= 0,9069
\end{equation}

\begin{equation}
ω_0= 0,9523
\end{equation}

\begin{equation}
\frac{ω_0}{Q}= 0,9957
\end{equation}

\begin{equation}
Q= \frac{ω_0}{0,9957}= \frac{0,9523}{0,9957}= 0,9564
\end{equation}

#####
##### Obtenemos $T_{HP}(s)$ en términos de $\omega_0$ y $Q$, para este caso particular:
#####

\begin{equation}
\boxed{T_{HP}(s)=\frac{0,8913*s^2} {s^2 + \frac{0,9523}{0,9564}*s + 0,9069}}
\end{equation}

#####
##### La $T_{HP}(s)$ anterior se trata de una transferencia normalizada y por lo general suele omitirse el correspondiente subíndice: lo correcto sería expresar dicha transferencia en términos de $s_N$.
##### Para desnormalizar la $T_{HP}(s)$ recordamos que $s_N= \frac{s}{\Omega_\omega}$ y siendo $\Omega_\omega= 2*\pi*4600 Hz$, obtenemos la transferencia del filtro pasa altos desnormalizada $T_{HP}(s)$:
#####

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * (\frac{s}{\Omega_\omega})^2 }{(\frac{s}{\Omega_\omega})^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*\frac{s}{\Omega_\omega} + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * \frac{1}{\Omega_\omega^2} * s^2 } { \frac{1}{\Omega_\omega^2} * s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}} * \frac{1}{\Omega_\omega} * s + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

#####
##### Dejamos mónico el polinomio denominador:
#####

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * \frac{\Omega_\omega^2}{\Omega_\omega^2} * s^2} { \frac{\Omega_\omega^2}{\Omega_\omega^2}*s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*\frac{\Omega_\omega^2}{\Omega_\omega}*s + \frac{2*ε}{\sqrt{ε^2+1}}*\Omega_\omega^2 }
\end{equation}

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * s^2} { s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}} * \Omega_\omega * s + \frac{2*ε}{\sqrt{ε^2+1}}*\Omega_\omega^2 }
\end{equation}

\begin{equation}
\boxed{T_{HP}(s)= \frac{0,8913*s^2} {s^2 +2,8777*10^4*s +7,5759*10^8}}
\end{equation}

```python
#-------------------------------------------------------------------------------------------------------------------#

# Inicializamos e importamos módulos

# Módulos externos
import matplotlib as mpl
import matplotlib.pyplot as plt
import math as m
import numpy as np
import scipy.signal as sig

fig_sz_x = 10
fig_sz_y = 9
fig_dpi = 150 # dpi

fig_font_size = 12

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
#plt.rcParams.update({'font.size':fig_font_size})

#-------------------------------------------------------------------------------------------------------------------#

# Importamos las funciones de PyTC2

from pytc2.sistemas_lineales import analyze_sys
from pytc2.sistemas_lineales import pretty_print_lti

#-------------------------------------------------------------------------------------------------------------------#

# Datos de plantilla

alfa_max = 1                        # dB
alfa_min = 20                       # dB

f_s_plantilla = 1.2e3               # Hz
f_p_plantilla = 4.6e3               # Hz

omega_p_plantilla = 2*m.pi*f_p_plantilla
norm_w = omega_p_plantilla
omega_p_plantilla /= norm_w
omega_p_pb_prototipo = 1/omega_p_plantilla

omega_s_plantilla = 2*m.pi*f_s_plantilla
omega_s_plantilla /= norm_w
omega_s_pb_prototipo = 1/omega_s_plantilla

print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- Datos de plantilla - Filtro pasa altos objetivo --')
print(' ')

print('α_min = {}'.format(alfa_min))
print('α_max = {}'.format(alfa_max))
print('fs = {}'.format(f_s_plantilla))
print('fp = {}'.format(f_p_plantilla))

print(' ')

print('ωp_n = {:3.4f}'.format(omega_p_plantilla))
print('ωs_n = {:3.4f}'.format(omega_s_plantilla))

print(' ')
print('-- Dominio transformado - Filtro pasa bajos prototipo --')
print(' ')

print('ωp = {:3.4f}'.format(omega_p_pb_prototipo))
print('ωs = {:3.4f}'.format(omega_s_pb_prototipo))

#-------------------------------------------------------------------------------------------------------------------#

# Calculamos e^2

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- ε: Grado de libertad de la funcion --')
print(' ')

print('ε = sqrt(10^(α_max/10)-1)')
eps = np.sqrt(10**(alfa_max/10)-1)

print(' ')

print('ε = {:3.4f}'.format(eps))
print('ε^2 = {:3.4f}'.format(eps**2))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Calculamos N iterando alfa_min

print(' ')
print('-- N: Orden del filtro --')
print(' ')
print('α_min = 10*log(1 + ε^2*C_n^2) con C_n^2= cosh^2(N*cosh^(-1)(ω_s))')
print(' ')

for N in range(1,10):
    C_nn = (np.cosh(N*(np.arccosh(omega_s_pb_prototipo))))**2
    att = 10*np.log10(1+(eps**2)*C_nn)    
    print('Para N= {:d}: ¿att = {:3.4f} dB >= α_min= {:d}?'.format(N, att, alfa_min))
    if att >= alfa_min:
        break

print(' ')

# Elegimos N de forma tal que α_min >= 20 dB:
print('N= {:d}'.format(N))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# De la resolución analítica, obtenemos |T(jw)|^2 |w=s/j = 1 / (1 + ε^2*C_n^2) = T(s)*T(-s) 

#Cn2 = 2w^2 - 1                 # Polinomio grado 2 de Chevyshev
Cn2 = [2, 0, -1]                # Coeficientes del polinomio grado 2 de Chevyshev
Cn4 = np.polymul(Cn2, Cn2)      # Elevamos al cuadrado el denominador
Cn4 = Cn4 * eps**2              # Multiplicamos por ε^2
Cn4[4]= Cn4[4] + 1              # Sumamos 1 al término independiente
Cn4[2]= -Cn4[2]                 # Cambiamos de signo al coeficiente del término cuadrático producto del reeemplazo de w^2= s^2/j^2
Cn4 = Cn4 / (4*eps**2)          # Hacemos mónico el polinomio

p_T_lp= np.roots(Cn4)

den_T_lp= np.poly([p_T_lp for p_T_lp in p_T_lp if p_T_lp.real < 0])
num_T_lp= [m.sqrt(1/(4*eps**2))]
#print(num_T_lp)
#print(den_T_lp)

num_T_hp, den_T_hp= sig.lp2hp(num_T_lp, den_T_lp)
#print(num_T_hp)
#print(den_T_hp)

print(' ')
print('-- z_i: Ceros obtenidos con roots(p) del polinomio numerador --')
print('-- obtenido de forma analítica --')
print(' ')

print('z1,2= {}'.format(np.roots(num_T_hp)))

print(' ')
print('-- p_i: Polos en el semiplano izquierdo obtenidos con roots(p) --')
print('-- del polinomio denominador obtenido de forma analítica --')
print(' ')

print('p1,2= {}'.format(np.roots(den_T_hp)))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos singularidades con la función de aproximación de Chevyshev del 
# filtro pasa bajos prototipo: cheb1ap(N, rp)

z, p, k = sig.cheb1ap(N, alfa_max)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos las singularidades que nos devuelve cheb1ap(N, rp) para obtener 
# el numerador y el denominador de la T_LP(s) del filtro pasa bajos prototipo 
# con zpk2tf(z,p,k):

num_lp, den_lp = sig.zpk2tf(z,p,k)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos la T_LP(s) del filtro pasa bajos prototipo para obtener 
# el numerador y el denominador de la T_HP(s) con lp2hp(num, den):

num_hp, den_hp = sig.lp2hp(num_lp, den_lp)

print(' ')
print('-- z_i: Ceros obtenidos con cheb1ap(N, rp) coincidentes --')
print('con los obtenidos con roots(p) de forma analítica --')
print(' ')

print('z1,2= {}'.format(np.roots(num_hp)))

print(' ')
print('-- p_i: Polos en el semiplano izquierdo obtenidos con cheb1ap(N, rp) --')
print('-- coincidentes con los obtenidos con roots(p) de forma analítica --')
print(' ')

print('p1,2= {}'.format(np.roots(den_hp)))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos (factorizamos) la T(s) en secciones de segundo orden (SOS) con la función tf2sos_analog()
# this_sos = tf2sos_analog(num_hp, den_hp)

t_T_hp= sig.TransferFunction(num_T_hp, den_T_hp)             # numerador y denominador resolución analítica

t_hp= sig.TransferFunction(num_hp, den_hp)                   # numerador y denominador con cheb1ap()

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('-- Función transferencia T(s) analítica NORMALIZADA: --')
print(' ')

pretty_print_lti(num_T_hp, den_T_hp)

print(' ')
print('-- Función transferencia T(s) con cheb1ap(N, rp) NORMALIZADA: --')
print(' ')

pretty_print_lti(num_hp, den_hp)

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

print('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --')
print(' ')

analyze_sys(t_hp, sys_name='Filtro pasa alto Chevyshev 2do orden')

#-------------------------------------------------------------------------------------------------------------------#
```

```python
#-------------------------------------------------------------------------------------------------------------------#

# Calculamos transferencia desnormalizada: producto de la desnormalización y de dejar mónico el polinomio
# denominador, los términos quedan afectados por los siguientes factores (Ver resolución analítica T_HP(s)
# desnormalizada)

num_T_hp_ = num_T_hp
den_T_hp_ = den_T_hp * [1, norm_w, norm_w**2]

t_T_hp_= sig.TransferFunction(num_T_hp_, den_T_hp_)

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('-- Función transferencia T(s) analítica DESNORMALIZADA: --')
print(' ')

pretty_print_lti(num_T_hp_, den_T_hp_)

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

print('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --')
print(' ')

analyze_sys(t_T_hp_, sys_name='Filtro pasa alto Chevyshev 2do orden')

#-------------------------------------------------------------------------------------------------------------------#
```
