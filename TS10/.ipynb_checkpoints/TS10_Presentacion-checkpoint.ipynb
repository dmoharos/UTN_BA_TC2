###
#### <u>Resolución analítica</u>
###
##### Comenzamos realizando una descripción de las singularidades de la función excitación $Z(s)$.
###
##### <u>Diagrama de singularidades de $Z(s)$</u>
###

<img src="TS10_Foster_paralelo_PZmap_Paso_1.png" style="margin-left:auto; margin-right:auto"/>

###
##### Como se puede observar, la función $Z(s)$ presenta:
###### -Un par de ceros imaginarios conjugados en $\sqrt{3}$,
###### -Un par de ceros imaginarios conjugados en $1$,
###### -Un polo simple en $0$ y
###### -Un par de polos imaginarios conjugados en $\sqrt{2}$
###
##### Aplicamos la inversión de la función $Z(s)$, obteniendo la función $Y(s) = \frac{ 1 }{ Z(s) }$:
###

\begin{equation*}
Y(s) = \frac{1}{Z(s)} = \frac{ s( s^2 +2 ) }{ ( s^2 +3 )( s^2 +1 ) } = \frac{ s^3 +2s }{ s^4 +4s^2 +3 }
\end{equation*}

###
##### Producto de la inversión, se reemplazan los ceros por polos y viceversa:
###
##### <u>Diagrama de singularidades de $Y(s)$</u>
###

<img src="TS10_Foster_paralelo_PZmap_Paso_2.png" style="margin-left:auto; margin-right:auto"/>

###
##### Aplicaremos el <u>teorema de Foster</u> para obtener los componentes del circuito en derivación:
###

\begin{equation*}
F= sk_{\infty} +\frac{ k_0 }{ s } +\sum\limits_{i=1}^{N} \frac{ 2k_is }{ s^2 +{\omega_i}^2 }  
\end{equation*}

###
##### Como se puede observar, $Y(s)$ posee ceros tanto en continua como en infinito, por lo tanto, los términos $\frac{ k_0 }{ s }$ y $sk_{\infty}$ son iguales a cero, obteniendo entonces:
###

\begin{equation*}
Y(s)= \sum\limits_{i=1}^{N} \frac{ 2k_is }{ s^2 +{\omega_i}^2 } = \frac{ 2k_1s }{ s^2 +3 } +\frac{ 2k_2s }{ s^2 +1 } 
\end{equation*}

###
##### Calculamos entonces los residuos $k_1$ y $k_2$:
###

\begin{equation*}
\lim_{{s^2 -> (-3)}} Y(s)= \frac{ 2k_1s }{ s^2 +3 } 
\end{equation*}

###

\begin{equation*}
2k_1 = \lim_{{s^2 -> (-3)}} \frac{ s^2 +3 }{ s } Y(s)
\end{equation*}

###

\begin{equation*}
\boxed{ 2k_1 = \lim_{{s^2 -> (-3)}} \frac{ s^2 +3 }{ s } \frac{ s( s^2 +2 ) }{ ( s^2 +3 )( s^2 +1 ) } = \frac{1}{2} }
\end{equation*}

###

\begin{equation*}
\lim_{{s^2 -> (-1)}} Y(s)= \frac{ 2k_2s }{ s^2 +1 } 
\end{equation*}

###

\begin{equation*}
2k_2 = \lim_{{s^2 -> (-1)}} \frac{ s^2 +1 }{ s } Y(s)
\end{equation*}

###

\begin{equation*}
\boxed{ 2k_2 = \lim_{{s^2 -> (-1)}} \frac{ s^2 +1 }{ s } \frac{ s( s^2 +2 ) }{ ( s^2 +3 )( s^2 +1 ) } = \frac{1}{2} }
\end{equation*}

###
##### Obtenidos $k_1$ y $k_2$, calculamos los componentes en cuestión sabiendo que el término $\sum\limits_{i=1}^{N} \frac{ 2k_is }{ s^2 +{\omega_i}^2 }$ representa cada una de las ramas L-C:
###

\begin{equation*}
\frac{ 1 }{ sL_i + \frac{1}{sC_i} } = \frac{ 2k_is }{ s^2 +{\omega_i}^2 }
\end{equation*}

###

\begin{equation*}
\frac{ 1 }{ sL_i + \frac{1}{sC_i} } = \frac{ 1 }{ \frac{ s^2 +{\omega_i}^2 }{ 2k_is } }
\end{equation*}

###

\begin{equation*}
\frac{ 1 }{ sL_i + \frac{1}{sC_i} } = \frac{ 1 }{ \frac{ s^2 }{ 2k_1s } + \frac{ {\omega_i}^2 }{ 2k_1s } }
\end{equation*}

###

\begin{equation*}
\frac{ 1 }{ sL_i + \frac{1}{sC_i} } = \frac{ 1 }{ \frac{ s }{ 2k_1 } + \frac{ 1 }{ \frac{ 2k_1s }{ {\omega_i}^2 }} }
\end{equation*}

###
##### Podemos observar entonces que los valores están dados por:
###

\begin{equation*}
L_i = \frac{ 1 }{ 2k_1 }
\end{equation*}

###

\begin{equation*}
C_i = \frac{ 2k_1 }{ {\omega_i}^2 }
\end{equation*}

###
##### Por caso:
###

\begin{equation*}
L_1 = \frac{ 1 }{ 2k_1 } = \frac{ 1 }{ \frac{1}{2} } = 2
\end{equation*}

###

\begin{equation*}
C_1 = \frac{ 2k_1 }{ {\omega_i}^2 } = \frac{ \frac{1}{2} }{ 3 } = \frac{1}{6} 
\end{equation*}

###

\begin{equation*}
L_2 = \frac{ 1 }{ 2k_2 } = \frac{ 1 }{ \frac{1}{2} } = 2
\end{equation*}

###

\begin{equation*}
C_2 = \frac{ 2k_2 }{ {\omega_i}^2 } = \frac{ \frac{1}{2} }{ 1 } = \frac{1}{2} 
\end{equation*}

###
##### Obtenemos finalmente el siguiente circuito:
###

<img src="TS10_Foster_paralelo_Circuito.png" style="margin-left:auto; margin-right:auto"/>

***

```python
#---------------------------------------------------------------------------------------------------------------------#
#-- Resolución simbólica --
#---------------------------------------------------------------------------------------------------------------------#

import sympy as sp
from pytc2.sintesis_dipolo import foster
from pytc2.dibujar import dibujar_foster_derivacion
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
from IPython.display import display, Markdown

#---------------------------------------------------------------------------------------------------------------------#

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
FF = ( s**3 + 2*s ) / ( s**4 + 4*s**2 +3 )

display( Markdown( '###' ) )
print_subtitle( '<u>Función de excitación</u>' )
print_latex( a_equal_b_latex_s( 'Y(s)', FF ) )

display( Markdown( '###' ) )
print_subtitle( '<u>Expansión según Teorema de Foster</u>' )
display( Markdown( '###' ) )
k0, koo, ki_wi, _, FF_foster = foster( FF )

print_latex( 'Y(s) = sk_{\infty} +\\frac{ k_0 }{ s } +\sum\limits_{i=1}^{N} \\frac{ 2k_is }{ s^2 +{\omega_i}^2 }' )
display( Markdown( '###' ) )

print_latex( a_equal_b_latex_s( 'k_0', k0 ) )
display( Markdown( '###' ) )

print_latex( a_equal_b_latex_s( r'2k_1 ', 1/ki_wi[0][1] ) )
display( Markdown( '###' ) )

print_latex( a_equal_b_latex_s( r'2k_2 ', 1/ki_wi[1][1] ) )
display( Markdown( '###' ) )

print_latex( a_equal_b_latex_s( 'k_\infty', koo ) )
display( Markdown( '###' ) )

print_latex( a_equal_b_latex_s( a_equal_b_latex_s( 'Y(s)', FF )[1:-1], FF_foster ) )
display( Markdown( '###' ) )

print_subtitle( '<u>Foster derivación</u>' )
display( Markdown( '###' ) )

dibujar_foster_derivacion( k0, koo, ki_wi, y_exc = FF )

#---------------------------------------------------------------------------------------------------------------------#
```

###
##### Reescribimos la función excitación $Z(s)$: 
###

\begin{equation*}
Z(s) = \frac{1}{Y(s)} = \frac{ s^4 +4s^2 +3 }{ s^3 + 2s }
\end{equation*}


###
#### <u>Método de Cauer 1:</u> remoción en $\infty$
###
##### <u>Resolución mediante divisiones sucesivas</u>
###

<img src="TS10_Cauer_1.png" style="margin-left:auto; margin-right:auto"/>

###
##### La función excitación $Z(s)$ según el método puede representarse de la siguiente forma:
###

\begin{equation*}
Z(s) = k_{\infty, 1} \cdot s + \frac{ 1 }{  k_{\infty, 2} \cdot s + \frac{ 1 }{  k_{\infty, 3} \cdot s + \frac{ 1 }{  k_{\infty, 4} \cdot s } } }
     = 1 \cdot s + \frac{ 1 }{  \frac{ 1 }{ 2 } \cdot s + \frac{ 1 }{  4 \cdot s + \frac{ 1 }{  \frac{ 1 }{ 6 } \cdot s } } } 
\end{equation*}

###
##### Obtenemos entonces el siguiente circuito:
###

<img src="TS10_Cauer_1_Circuito.png" style="margin-left:auto; margin-right:auto"/>

###
##### Notar que en las sucesivas divisiones, el nuevo dividendo es el divisor anterior y que el nuevo divisor es el resto de la división anterior.
###
##### Este método nos permite obtener el valor del componente que se remueve y la inmitancia resultante de la remoción: el primero de ellos dado por el resultado del cociente de los polinomios y el segundo como la división entre el resto y el divisor.
###
##### Procedemos a realizar la verificación del método:
###

\begin{equation*}
Z_2(s) = Z(s) - k_{\infty}s
\end{equation*}

###
##### De igual manera, podemos escribir:
###

\begin{equation*}
Z(s) = k_{\infty}s +Z_2(s)
\end{equation*}

###
##### Por caso, en primera remoción, obtenemos que:
###

\begin{equation*}
Z(s) = \frac{ s^4 +4s^2 +3 }{ s^3 + 2s } = s +\frac{ 2s^2 +3 }{ s^3 + 2s }
\end{equation*}

##### Operando de igual manera, pueden verificarse las restantes remociones.

```python
#---------------------------------------------------------------------------------------------------------------------#
#-- Resolución simbólica / numérica --
#---------------------------------------------------------------------------------------------------------------------#

import sympy as sp
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
from IPython.display import display, Markdown
from pytc2.sintesis_dipolo import cauer_LC
from pytc2.dibujar import dibujar_cauer_LC

#---------------------------------------------------------------------------------------------------------------------#

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
FF = ( s**4 + 4*s**2 + 3 )/( s**3 + 2*s )

display( Markdown( '###' ) )
print_subtitle( '<u>Función de excitación</u>' )
print_latex(a_equal_b_latex_s('Z(s)', FF))
display( Markdown( '###' ) )

# Implementaremos FF mediante Cauer 1 o remociones continuas en infinito
koo, F_cauer_oo, rem = cauer_LC(FF, remover_en_inf=True)

display( Markdown( '###' ) )
print_subtitle( '<u>Expansión según Método de Cauer 1: remociones en infinito</u>' )
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('Z(s)', FF)[1:-1], F_cauer_oo ))
display( Markdown( '###' ) )

# Tratamos a nuestra función inmitancia como una Z
dibujar_cauer_LC(koo, z_exc = F_cauer_oo)

#---------------------------------------------------------------------------------------------------------------------#
```

     = \frac{ 3 }{ 2 \cdot s } + \frac{ 1 }{ \frac{ 4 }{ 5 \cdot s } + \frac{ 1 }{ \frac{ 25 }{ 2 \cdot s } + \frac{ 1 }{ \frac{ 1 }{ 5 \cdot s } } } }

```python
#---------------------------------------------------------------------------------------------------------------------#
#-- Resolución simbólica / numérica --
#---------------------------------------------------------------------------------------------------------------------#

import sympy as sp
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
from IPython.display import display, Markdown
from pytc2.sintesis_dipolo import cauer_LC
from pytc2.dibujar import dibujar_cauer_LC

#---------------------------------------------------------------------------------------------------------------------#

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
FF = ( s**4 + 4*s**2 + 3 )/( s**3 + 2*s )

display( Markdown( '###' ) )
print_subtitle( '<u>Función de excitación</u>' )
print_latex(a_equal_b_latex_s('Z(s)', FF))
display( Markdown( '###' ) )

# Implementaremos FF mediante Cauer 2 o remociones continuas en cero
k0, F_cauer_0, rem = cauer_LC(FF, remover_en_inf=False)

display( Markdown( '###' ) )
print_subtitle( '<u>Expansión según Método de Cauer 2: remociones en continua</u>' )
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('Z(s)', FF)[1:-1], F_cauer_0 ))
display( Markdown( '###' ) )

# Tratamos a nuestra función inmitancia como una Z
dibujar_cauer_LC(k0, z_exc = F_cauer_0)

#---------------------------------------------------------------------------------------------------------------------#
```

###
##### En primera instancia, invertimos la función de excitación $Y(s)$ para obtener la correspondiente $Z(s)$: a continuación explicaremos los motivos.
###

\begin{equation*}
Z(s) = \frac{1}{Y(s)} = \frac{ ( s^2 +2 )( s^2 +5 ) }{ 3s( s^2 +\frac{7}{3} ) }
\end{equation*}

###
##### <u>Diagrama de singularidades de $Z(s)$</u>
###

<img src="TS10_Remocion_Parcial_PZmap_Paso_2.png" style="margin-left:auto; margin-right:auto"/>

###
##### Podemos observar que presenta un polo en continua que nos permitirá sintonizar el cero a la frecuencia de resonancia solicitada del tanque LC en derivación, mediante la remoción parcial de su residuo. 
###

<img src="TS10_Remocion_Parcial_PZmap_Paso_3.png" style="margin-left:auto; margin-right:auto"/>

###
##### Obtenemos el residuo de forma analítica:
###

\begin{equation*}
Z_{21}(s) = Z(s) - Z_1(s) = Z(s) - \frac{ k_0' }{ s }
\end{equation*}

###
##### En la frecuencia de interés, sucede que:
###

\begin{equation*}
\lim_{{s^2 -> (-1)}} Z_{21}(s) = 0
\end{equation*}

###
##### Operamos algebraicamente y obtenemos $k_0'$ como:
###

\begin{equation*}
k_0' = \lim_{{s^2 -> (-1)}} \left[ Z(s) \cdot s \right]
\end{equation*}

###

\begin{equation*}
k_0' = \lim_{{s^2 -> (-1)}} \left[ \frac{ ( s^2 +2 )( s^2 +5 ) }{ 3s( s^2 +\frac{7}{3} ) } \cdot s \right] 
\end{equation*}

###

\begin{equation*}
k_0' = \frac{ ( -1 +2 )( -1 +5 ) }{ 3( -1 +\frac{7}{3} ) } = 1
\end{equation*}

###
##### Los valores de $Z_1(s)$ y $Z_2(s)$ serán entonces:
###

\begin{equation*}
\boxed{ Z_1(s) = \frac{k_0'}{s} = \frac{1}{s} }
\end{equation*}

###
##### Este residuo representa un capacitor en serie de valor unitario.
###

\begin{equation*}
Z_2(s) = Z(s) - Z_1(s) = \frac{ ( s^2 +2 )( s^2 +5 ) }{ 3s( s^2 +\frac{7}{3} ) } - \frac{1}{s}
\end{equation*}

###

\begin{equation*}
Z_2(s) = \frac{ ( s^2 +2 )( s^2 +5 ) - 3( s^2 +\frac{7}{3} ) }{ 3s( s^2 +\frac{7}{3} ) }
\end{equation*}

###

\begin{equation*}
Z_2(s) = \frac{ s^4 + 7s^2 + 10 - 3s^2 -7 }{ 3s^3 +7s }
\end{equation*}

###

\begin{equation*}
\boxed{ Z_2(s) = \frac{ s^4 + 4s^2 + 3 }{ 3s^3 +7s } = \frac{ ( s^2 + 1 )( s^2 + 3 ) }{ 3s( s^2 +\frac{7}{3} ) } }
\end{equation*}

###
##### <u>Verificación de la resonancia de los tanques LC en derivación</u>
###

\begin{equation*}
Y_2(s) = \frac{ 1 }{ Z_2(s) } = \frac{ 3s( s^2 +\frac{7}{3} ) }{ ( s^2 + 1 )( s^2 + 3 ) }
\end{equation*}

###
##### <u>Diagrama de singularidades de $Y_2(s)$</u>
###

<img src="TS10_Remocion_Parcial_PZmap_Paso_4.png" style="margin-left:auto; margin-right:auto"/>

###
##### Como se puede observar, $Y_2(s)$ posee ceros tanto en continua como en infinito, por lo tanto, los términos $\frac{ k_0 }{ s }$ y $sk_{\infty}$ son iguales a cero, lo que nos permite plantear el teorema de Foster en derivación conociendo la topología del circuito.
###

\begin{equation*}
Y_2(s)= \sum\limits_{i=1}^{N} \frac{ 2k_is }{ s^2 +{\omega_i}^2 } = \frac{ 2k_2s }{ s^2 +1 } +\frac{ 2k_3s }{ s^2 +3 } 
\end{equation*}

###
##### Calculamos entonces los residuos $k_2$ y $k_3$:
###

\begin{equation*}
\lim_{{s^2 -> (-1)}} Y_2(s)= \frac{ 2k_2s }{ s^2 +1 } 
\end{equation*}

###

\begin{equation*}
2k_2 = \lim_{{s^2 -> (-1)}} \frac{ s^2 +1 }{ s } Y_2(s)
\end{equation*}

###

\begin{equation*}
2k_2 = \lim_{{s^2 -> (-1)}} \frac{ s^2 +1 }{ s } \frac{ 3s( s^2 +\frac{7}{3} ) }{ ( s^2 + 1 )( s^2 + 3 ) }
\end{equation*}

###

\begin{equation*}
\boxed{ 2k_2 = \lim_{{s^2 -> (-1)}} \frac{ 3 ( s^2 +\frac{7}{3} ) }{ s^2 + 3 } = 2 }
\end{equation*}

###

\begin{equation*}
\lim_{{s^2 -> (-3)}} Y_2(s)= \frac{ 2k_3s }{ s^2 +3 } 
\end{equation*}

###

\begin{equation*}
2k_3 = \lim_{{s^2 -> (-3)}} \frac{ s^2 +3 }{ s } Y_2(s)
\end{equation*}

###

\begin{equation*}
2k_3 = \lim_{{s^2 -> (-3)}} \frac{ s^2 +3 }{ s } \frac{ 3s( s^2 +\frac{7}{3} ) }{ ( s^2 + 1 )( s^2 + 3 ) }
\end{equation*}

###

\begin{equation*}
\boxed{ 2k_3 = \lim_{{s^2 -> (-3)}} \frac{ 3 ( s^2 +\frac{7}{3} ) }{ s^2 + 1 } = 1 }
\end{equation*}

###
##### En función de los residuos obtenidos, calculamos los componentes de los tanques LC en derivación para posteriormente verificar las frecuencias de resonancia:
###

\begin{equation*}
Y_2(s)= \sum\limits_{i=1}^{N} \frac{ 2k_is }{ s^2 +{\omega_i}^2 } = \sum\limits_{i=1}^{N} \frac{ 1 }{ \frac{ s^2 + {\omega_i}^2 }{ 2k_is } } = \sum\limits_{i=1}^{N} \frac{ 1 }{ \frac{ s^2 }{ 2k_is } + \frac{ {\omega_i}^2 }{ 2k_is } } = \sum\limits_{i=1}^{N} \frac{ 1 }{ \frac{ s }{ 2k_i } + \frac{1}{ \frac{ 2k_is }{ {\omega_i}^2 } } } = \frac{ 1 }{ \frac{ s }{ 2k_2 } + \frac{1}{ \frac{ 2k_2s }{ {\omega_2}^2 } } } + \frac{ 1 }{ \frac{ s }{ 2k_3 } + \frac{1}{ \frac{ 2k_3s }{ {\omega_3}^2 } } }
\end{equation*}

###
##### Por caso:
###

\begin{equation*}
L_2 = \frac{ 1 }{ 2k_2 } = \frac{ 1 }{ 2 }
\end{equation*}

###

\begin{equation*}
C_2 = \frac{ 2k_2 }{ {\omega_2}^2 } = \frac{ 2 }{ 1 } = 2
\end{equation*}

###

\begin{equation*}
L_3 = \frac{ 1 }{ 2k_3 } = \frac{ 1 }{ 1 } = 1
\end{equation*}

###

\begin{equation*}
C_3 = \frac{ 2k_3 }{ {\omega_3}^2 } = \frac{ 1 }{ 3 }
\end{equation*}

###
##### Las frecuencias de resonancia están dadas por:
###

\begin{equation*}
{\omega_i}^2 = \frac{ 1 }{ L_iC_i }
\end{equation*}

###
##### Por caso:
###

\begin{equation*}
{\omega_2}^2 = \frac{ 1 }{ L_2C_2 } = \frac{ 1 }{ \frac{1}{2} 2 } = 1
\end{equation*}

###

\begin{equation*}
{\omega_3}^2 = \frac{ 1 }{ L_3C_3 } = \frac{ 1 }{ 1\frac{1}{3} } = 3
\end{equation*}

###
##### Se verifica que el tanque en derivación $L_2C_2$ resuena a $1\frac{rad}{s}$.
###
##### Otra alternativa, que es la que utilizaremos a continuación para verificar de forma simbólica/numérica los resultados obtenidos, es realizar la síntesis completa del circuito propuesto mediante remociones parciales y totales.
###

```python
#---------------------------------------------------------------------------------------------------------------------#
#-- Resolución simbólica --
#---------------------------------------------------------------------------------------------------------------------#

import numpy            as     np
import sympy            as     sp
from   pytc2.remociones import remover_polo_dc, remover_polo_jw
from   pytc2.dibujar    import display, dibujar_puerto_entrada, dibujar_funcion_exc_abajo, dibujar_elemento_serie 
from   pytc2.dibujar    import dibujar_tanque_derivacion, dibujar_espacio_derivacion, Capacitor
from   pytc2.general    import print_latex, print_subtitle, a_equal_b_latex_s
from   IPython.display  import display, Markdown

#---------------------------------------------------------------------------------------------------------------------#

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
YY = 3*s*(s**2+sp.Rational(7,3))/(s**2+2)/(s**2+5)

#---------------------------------------------------------------------------------------------------------------------#

# Dibujamos el circuito genérico
display( Markdown( '###' ) )
print_subtitle('<u>Circuito propuesto</u>')
display( Markdown( '###' ) )
d = dibujar_puerto_entrada('', voltage_lbl = ('+', '$V$', '-'), current_lbl = '$I$')
d, zz_lbl = dibujar_funcion_exc_abajo(d, 'Y', YY, hacia_salida = True, k_gap_width = 0.5)
d = dibujar_elemento_serie(d, Capacitor, 'C1')
d = dibujar_tanque_derivacion(d, 'L2', 'C2')
d = dibujar_espacio_derivacion(d)
d = dibujar_tanque_derivacion(d, 'L3', 'C3')
display(d)

#---------------------------------------------------------------------------------------------------------------------#
display( Markdown( '###' ) )
print_subtitle('<u>Admitancia de entrada a la red</u>')
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Y(s)', YY))
display( Markdown( '###' ) )

#---------------------------------------------------------------------------------------------------------------------#

# Restricción circuital: L2*C2 = 1 r/s
# Remoción parcial en DC de 1/YY

omega_L2C2 = 1

Z2, Zc1 = remover_polo_dc( 1/YY, omega_zero = omega_L2C2 )

# Yc1 es la admitancia removida
# Extraemos C1
C1 = 1/(s*Zc1)

display( Markdown( '###' ) )
print_subtitle('<u>Remoción parcial</u>')
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Z_1(s) = \\frac{k^p_0}{s}', Zc1))
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Z_2(s)', Z2))
display( Markdown( '###' ) )

#---------------------------------------------------------------------------------------------------------------------#

Y4, Yt2, L2, C2 = remover_polo_jw( 1/Z2, isImpedance = False, omega = omega_L2C2 )

display( Markdown( '###' ) )
print_subtitle('<u>Remoción total tanque en derivación $L_2C_2$</u>')
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Y_3(s)', Yt2))
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Y_4(s)', Y4))
display( Markdown( '###' ) )

#---------------------------------------------------------------------------------------------------------------------#

# quedaría solo un tanque en Y4, no especifico omega.
Y6, Yt3, L3, C3 = remover_polo_jw( Y4, isImpedance = False )

display( Markdown( '###' ) )
print_subtitle('<u>Remoción total tanque en derivación $L_3C_3$</u>')
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Y_5(s)', Yt3))
display( Markdown( '###' ) )
print_latex(a_equal_b_latex_s('Y_6(s)', Y6))
display( Markdown( '###' ) )

#---------------------------------------------------------------------------------------------------------------------#

# Dibujamos la red resultante:

display( Markdown( '###' ) )
print_subtitle('<u>Circuito resultante</u>')
display( Markdown( '###' ) )
d = dibujar_puerto_entrada('', voltage_lbl = ('+', '$V$', '-'), current_lbl = '$I$')
d, zz_lbl = dibujar_funcion_exc_abajo( d, 'Y', YY, hacia_salida = True, k_gap_width = 0.5 )
d = dibujar_elemento_serie(d, Capacitor, C1)
d = dibujar_tanque_derivacion(d, L2, C2)
d = dibujar_espacio_derivacion(d)
d = dibujar_tanque_derivacion(d, L3, C3)
display(d)

#---------------------------------------------------------------------------------------------------------------------#
```
