```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

#------------------------------------------------------------------------------------------------------------------------------------#
#                                               -- Inicialización de librerías --                                                    #
#------------------------------------------------------------------------------------------------------------------------------------#

from scipy.signal import TransferFunction
from scipy import signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

#------------------------------------------------------------------------------------------------------------------------------------#
#                                         -- Configuración para visualización de gráficos --                                         #
#------------------------------------------------------------------------------------------------------------------------------------#

mpl.rcParams['figure.figsize'] = (10,10)
fig_sz_x= 14
fig_sz_y= 13
fig_dpi= 80 # dpi
fig_font_size= 10
plt.rcParams.update({'font.size':fig_font_size})

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                -- Declaración de variables --                                                      #
#------------------------------------------------------------------------------------------------------------------------------------#

#-- Valores de componentes --
R1= 10e3
R2= (np.sqrt(2)/2)*R3
R3= 10e3
R4= 10e3
C= 100e-6

#-- Norma de impedancia --
Ω_z= R3

#-- Valores de componentes normalizados--
R1_n= R1/Ω_z
R2_n= R2/Ω_z
R3_n= R3/Ω_z
R4_n= R4/Ω_z
C_n= C*Ω_z 

#-- Características de la transferencia --
k= 1/(R1*R3*C**2)
qq= R2/R3
ω_0= 1/(R3*C)

#-- Norma de pulsación --
Ω_ω= ω_0

#-- Características de la transferencia normalizadas--
ω_0_n= ω_0/Ω_ω

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                       -- Gráficos --                                                               #
#                       -- Transferencia (módulo y fase), diagramas de polos y ceros y retardo de grupo --                           #
#------------------------------------------------------------------------------------------------------------------------------------#

my_tf = TransferFunction( [-k], [1, ω_0/qq, ω_0**2] )

plt.close('all')

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )
pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq))
GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
    
#------------------------------------------------------------------------------------------------------------------------------------#
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

#------------------------------------------------------------------------------------------------------------------------------------#
#                                               -- Inicialización de librerías --                                                    #
#------------------------------------------------------------------------------------------------------------------------------------#

from scipy.signal import TransferFunction
from scipy import signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

#------------------------------------------------------------------------------------------------------------------------------------#
#                                         -- Configuración para visualización de gráficos --                                         #
#------------------------------------------------------------------------------------------------------------------------------------#

mpl.rcParams['figure.figsize'] = (10,10)
fig_sz_x= 14
fig_sz_y= 13
fig_dpi= 80 # dpi
fig_font_size= 10
plt.rcParams.update({'font.size':fig_font_size})

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                -- Declaración de variables --                                                      #
#------------------------------------------------------------------------------------------------------------------------------------#

#-- Valores de componentes --
R1= 10e3
R2= (np.sqrt(2)/2)*R3
R3= 10e3
R4= 10e3
C= 100e-6

#-- Norma de impedancia --
Ω_z= R3

#-- Valores de componentes normalizados--
R1_n= R1/Ω_z
R2_n= R2/Ω_z
R3_n= R3/Ω_z
R4_n= R4/Ω_z
C_n= C*Ω_z 

#-- Características de la transferencia --
ω_0= 1/(R3*C)
qq= ω_0/np.sqrt(2)

#-- Norma de pulsación --
Ω_ω= ω_0

#-- Características de la transferencia normalizadas--
ω_0_n= ω_0/Ω_ω

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                       -- Gráficos --                                                               #
#                       -- Transferencia (módulo y fase), diagramas de polos y ceros y retardo de grupo --                           #
#------------------------------------------------------------------------------------------------------------------------------------#

my_tf = TransferFunction( [1], [1, ω_0/qq, ω_0**2] )

plt.close('all')

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )
pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq))
GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
    
#------------------------------------------------------------------------------------------------------------------------------------#
```

#### Simulación de transferencia (módulo y fase), diagrama de polos y ceros y retardo de grupo

####

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

#------------------------------------------------------------------------------------------------------------------------------------#
#                                               -- Inicialización de librerías --                                                    #
#------------------------------------------------------------------------------------------------------------------------------------#

from scipy.signal import TransferFunction
from scipy import signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

#------------------------------------------------------------------------------------------------------------------------------------#
#                                         -- Configuración para visualización de gráficos --                                         #
#------------------------------------------------------------------------------------------------------------------------------------#

mpl.rcParams['figure.figsize'] = (10,10)
fig_sz_x= 14
fig_sz_y= 13
fig_dpi= 80 # dpi
fig_font_size= 10
plt.rcParams.update({'font.size':fig_font_size})

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                -- Declaración de variables --                                                      #
#------------------------------------------------------------------------------------------------------------------------------------#

#-- Valores de componentes --
R1= 10e3
R2= (np.sqrt(2)/2)*R3
R3= 10e3
R4= 10e3
C= 100e-6

#-- Norma de impedancia --
Ω_z= R3

#-- Valores de componentes normalizados--
R1_n= R1/Ω_z
R2_n= R2/Ω_z
R3_n= R3/Ω_z
R4_n= R4/Ω_z
C_n= C*Ω_z 

#-- Características de la transferencia --
k= 1/(R1*C)
ω_0= 1/(R3*C)
qq= R2/R3

#-- Norma de pulsación --
Ω_ω= ω_0

#-- Características de la transferencia normalizadas--
ω_0_n= ω_0/Ω_ω
qq_n= qq/Ω_ω

#------------------------------------------------------------------------------------------------------------------------------------#
#                                                       -- Gráficos --                                                               #
#                       -- Transferencia (módulo y fase), diagramas de polos y ceros y retardo de grupo --                           #
#------------------------------------------------------------------------------------------------------------------------------------#

my_tf = TransferFunction( [-k, 0], [1, ω_0_n/qq_n, ω_0_n**2] )

plt.close('all')

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq_n) )
pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq_n))
GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq_n))
    
#------------------------------------------------------------------------------------------------------------------------------------#
```
