#### <span style="text-decoration: underline;">Cálculo del grado de libertad de la función $ε$ y el orden del filtro N</span>

#####
##### Tomando como referencias las ecuaciones de $|T(jω)|^2$ en su aproximación de Chevyshev [$^{(1)}$](#1) [$^{(2)}$](#2):
#####

\begin{equation*}
|T(jΩ)|^2= T(jΩ)*T(-jΩ)= {T(s)*T(-s)}|_{s= jΩ}= \frac{1}{1+ε^2*C_n^2(Ω)}
\end{equation*}

#####
##### Podemos entonces establecer la atenuación en función de la ecuación anterior como la inversa de $|T(jΩ)|^2$:
#####

\begin{equation*}
|α|^2= \frac {1}{|T(jΩ)|^2}= \frac{1}{\frac{1}{1+ε^2*C_n^2(Ω)}}= 1+ε^2*C_n^2(Ω)
\end{equation*}

#####
##### Considerando que $|α|_{dB} \triangleq α_{dB}$, podemos expresar la ecuación en dB [$^{(3)}$](#3):
#####

\begin{equation*}
α_{dB}= 10*log[1+ε^2*C_n^2(Ω)]
\end{equation*}

#####
##### Despejando $ε^2$ de la ecuación anterior, obtenemos [$^{(4)}$](#4):
#####

\begin{equation*}
ε^2= \frac{10^{\frac{α_{máx}}{10}}-1}{C_n^2(Ω)}
\end{equation*}

#####
##### Considerando la máxima atenuación requerida por la plantilla $α_{máx}$ en el vértice de fin de la banda de paso $Ω_{p_n}$, obtenemos $ε^2$:
#####

\begin{equation*}
ε^2= \frac{10^{\frac{α_{máx}}{10}}-1}{C_n^2(Ω_{p_n})}= \frac{10^{\frac{20}{10}}-1}{1}
\end{equation*}

\begin{equation*}
\boxed{ε^2= 0,2589}
\end{equation*}

\begin{equation*}
\boxed{ε= 0,5088}
\end{equation*}

#####
##### Considerando la mínima atenuación requerida por la plantilla $α_{mín}$ en el vértice del inicio de la banda de detenida $Ω_{s_n}$, iteramos $N$ de forma tal que la atenuación sea mayor o igual a obtenemos $α_{mín}$:
#####

\begin{equation*}
α_{mín} = 10*log[1 + ε^2*C_n^2(Ω_{s_n})]
\end{equation*}

\begin{equation*}

\begin{equation*}

\begin{equation*}
\boxed{N= 2}
\end{equation*}

#####
#### <span style="text-decoration: underline;">Cálculo de la función transferencia $T_{HP}(s)$</span>
#####

##### Recordando que $|T(jΩ)|^2= T(jΩ)*T(-jΩ)= {T(s)*T(-s)}|_{s= jΩ}= \frac{1}{1+ε^2*C_n^2(Ω)}$ y habiendo obtenido $ε^2$ y N, procedemos a calcular los polinomios de orden N de Chevyshev definidos como:
#####

\begin{equation*}
C_n^2= cosh^2(n*cosh^{-1}(Ω_{s}))
\end{equation*}

#####
##### Utilizando identidades trigonómetricas [$^{(5)}$](#5), calculamos el polinomio para el grado N obtenido:
#####

\begin{equation*}
C_0(Ω)= 1 
\end{equation*}

\begin{equation*}
C_1(Ω)= Ω 
\end{equation*}

\begin{equation*}
C_n(Ω)= 2*Ω*C_{n-1}(Ω) - C_{n-2}(Ω) 
\end{equation*}

\begin{equation*}
C_2(Ω)= 2*Ω*Ω - 1= 2*Ω^2-1 
\end{equation*}

#####
##### Reemplazando el polinomio de grado 2 en la ecuación de $|T(jΩ)|^2= \frac{1}{1+ε^2*C_n^2(Ω)}$ y operando algebraicamente, se obtiene la transferencia del filtro pasa bajos prototipo normalizada $T_{LP}(\$)$:
#####

\begin{equation*}
\boxed{T_{LP}($)=\frac{\frac{1}{2*ε}}{$^2 + \sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*$ + \frac{\sqrt{ε^2+1}}{2*ε}}= \frac{0.9827}{$^2 + 1,0978*$ + 1,1026}}
\end{equation*}

#####
##### Aplicando el núcleo de la transformación $K(s)= \$= \frac{1}{s}$, se obtener la transferencia pasa altos objetivo normalizada $T_{HP}(s)$:
#####

\begin{equation*}
\boxed{T_{HP}(s)= \frac {0,8913*s^2} {s^2 +0,9957*s +0,9069}}
\end{equation*}

#####
##### En términos de $\omega_0$ y $Q$, podemos expresarla de la siguiente manera:
#####

\begin{equation*}
\boxed{T_{HP}(s)=\frac{0,8913*s^2} {s^2 + \frac{0,9523}{0,9564}*s + 0,9069}}
\end{equation*}

#####
##### Desnormalizando, se obtiene:
##### 

\begin{equation*}
\boxed{T_{HP}(s)= \frac{0,8913*s^2} {s^2 +2,8777*10^4*s +7,5759*10^8}}
\end{equation*}

#####

#####

```python
# ------------------------------------------------------------------------------------------------- #
# Librerias utilizadas
# ------------------------------------------------------------------------------------------------- #

import scipy.signal as sig
import sympy as sp
from sympy import N
import numpy as np
from numpy import pi

import matplotlib as mpl
from IPython.display import display, Markdown, Math

#from pytc2.sistemas_lineales import tf2sos_analog
from pytc2.sistemas_lineales import pretty_print_lti
from pytc2.sistemas_lineales import pretty_print_SOS
#from pytc2.sistemas_lineales import pretty_print_bicuad_omegayq
from pytc2.sistemas_lineales import analyze_sys
from pytc2.general import print_latex
#from pytc2.general import print_subtitle

# ------------------------------------------------------------------------------------------------- #
```

```python
# ------------------------------------------------------------------------------------------------- #
# Ajustes de visualización
# ------------------------------------------------------------------------------------------------- #

fig_sz_x = 10
fig_sz_y = 9
fig_dpi = 150 # dpi

fig_font_size = 12

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
#plt.rcParams.update({'font.size':fig_font_size})

# ------------------------------------------------------------------------------------------------- #
```

```python
# ------------------------------------------------------------------------------------------------- #
# Datos de plantilla
# ------------------------------------------------------------------------------------------------- #

alfa_max = 1                                    # dB
alfa_min = 20                                   # dB

f_s_plantilla = 1.2e3                           # Hz
f_p_plantilla = 4.6e3                           # Hz

omega_p_plantilla = 2*pi*f_p_plantilla
Omega_omega= omega_p_plantilla
omega_p_plantilla /= Omega_omega
omega_p_pb_prototipo = 1/omega_p_plantilla

omega_s_plantilla = 2*pi*f_s_plantilla
omega_s_plantilla /= Omega_omega
omega_s_pb_prototipo = 1/omega_s_plantilla

display(Markdown('***'))
display(Markdown('#####'))
display(Markdown('##### -- Datos de plantilla - Filtro pasa altos objetivo --'))
display(Markdown('#####'))

display(Math(r'\alpha_{min}= ' + sp.latex(alfa_min)))
display(Math(r'\alpha_{max}= ' + sp.latex(alfa_max)))
display(Math(r'fs= ' + sp.latex(f_s_plantilla)))
display(Math(r'fp= ' + sp.latex(f_p_plantilla)))

display(Markdown('#####'))
display(Markdown('##### -- Dominio transformado - Filtro pasa bajos prototipo --'))
display(Markdown('#####'))

display(Math(r'\omega_{p_{n}}= ' + sp.latex(N(omega_p_plantilla, 4))))
display(Math(r'\omega_{s_{n}}= ' + sp.latex(N(omega_s_plantilla, 4))))

display(Markdown('#####'))
display(Markdown('***'))

# ------------------------------------------------------------------------------------------------- #
```

```python
# ------------------------------------------------------------------------------------------------- #
# Calculamos e^2
# ------------------------------------------------------------------------------------------------  #

display(Markdown('***'))
display(Markdown('#####'))
display(Markdown('##### -- $\epsilon$: Grado de libertad de la funcion --'))
display(Markdown('#####'))

display(Math(r'\epsilon = \sqrt{10^\frac{\alpha_{max}}{10}-1}'))
eps = np.sqrt(10**(alfa_max/10)-1)

display(Math(r'\epsilon= ' + sp.latex(N(eps, 4))))
display(Math(r'\epsilon^2= ' + sp.latex(N(eps**2, 4))))

display(Markdown('#####'))
display(Markdown('***'))

# ------------------------------------------------------------------------------------------------- #
```

```python
# -------------------------------------------------------------------------------------------------- #
# Calculamos N iterando alfa_min
# -------------------------------------------------------------------------------------------------- #

display(Markdown('***'))
display(Markdown('#####'))
display(Markdown('-- $N$: Orden del filtro --'))
display(Markdown('#####'))
display(Math(r'\alpha_{min} = 10*log(1 + \epsilon^2*C_n^2) \quad ; \quad C_n^2= cosh^2\left[N*cosh^{-1}(\omega_s)\right]'))

for nn in range(1,10):
    C_nn = (np.cosh(nn*(np.arccosh(omega_s_pb_prototipo))))**2
    att = 10*np.log10(1+(eps**2)*C_nn)    
    display(Math(r'Para \quad N= ' + sp.latex(nn) + r': \quad ¿att= ' + sp.latex(N(att, 4)) + r'dB >= \alpha_{min}= ' + sp.latex(N(alfa_min, 4)) + r'?'))
    if att >= alfa_min:
        break
        
# Elegimos N de forma tal que α_min >= 20 dB:

display(Markdown('#####'))
display(Math('N= ' + sp.latex(nn)))
display(Markdown('#####'))
display(Markdown('***'))

# -------------------------------------------------------------------------------------------------- #
```

```python
# ------------------------------------------------------------------------------------------------- #
# De la resolución analítica, obtenemos |T(jw)|^2 |w=s/j = 1 / (1 + ε^2*C_n^2) = T(s)*T(-s) 
# ------------------------------------------------------------------------------------------------- #

#Cn2 = 2w^2 - 1                 # Polinomio grado 2 de Chevyshev
Cn2 = [2, 0, -1]                # Coeficientes del polinomio grado 2 de Chevyshev
Cn4 = np.polymul(Cn2, Cn2)      # Elevamos al cuadrado el denominador
Cn4 = Cn4 * eps**2              # Multiplicamos por ε^2
Cn4[4]= Cn4[4] + 1              # Sumamos 1 al término independiente
Cn4[2]= -Cn4[2]                 # Cambiamos de signo al coeficiente del término cuadrático producto 
                                # del reeemplazo de w^2= s^2/j^2
Cn4 = Cn4 / (4*eps**2)          # Hacemos mónico el polinomio

# Obtenemos las raíces de Cn4 que se corresponden con los polos de |T(jw)|^2 |w=s/j = T(s)*T(-s):
p_T_lp= np.roots(Cn4)

# Seleccionamos aquellas raíces que se encuentren en el semiplano izquierdo correspondientes a T(s):
den_T_lp= np.poly([p_T_lp for p_T_lp in p_T_lp if p_T_lp.real < 0])

# Obtenemos el numerador de T(s):
num_T_lp= [np.sqrt(1/(4*eps**2))]

# Aplicamos el núcleo de la transformación pasa altos a través de la función lp2hp():
num_T_hp, den_T_hp= sig.lp2hp(num_T_lp, den_T_lp)

display(Markdown('#####'))
display(Markdown('-- z_i: Ceros obtenidos con roots(p) del polinomio numerador obtenido de forma analítica --'))
display(Markdown('#####'))

display(Math(r'z_{1,2}= ' + sp.latex(np.roots(num_T_hp))))

display(Markdown('#####'))
display(Markdown('-- p_i: Polos en el semiplano izquierdo obtenidos con roots(p) del polinomio denominador obtenido de forma analítica --'))
display(Markdown('#####'))

display(Math(r'p_{1,2}= ' + sp.latex(np.roots(den_T_hp))))

display(Markdown('#####'))
display(Markdown('***'))
```

```python
#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos singularidades con la función de aproximación de Chevyshev del 
# filtro pasa bajos prototipo: cheb1ap(N, rp)

z, p, k = sig.cheb1ap(nn, alfa_max)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos las singularidades que nos devuelve cheb1ap(N, rp) para obtener 
# el numerador y el denominador de la T_LP(s) del filtro pasa bajos prototipo 
# con zpk2tf(z,p,k):

num_lp, den_lp = sig.zpk2tf(z,p,k)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos la T_LP(s) del filtro pasa bajos prototipo para obtener 
# el numerador y el denominador de la T_HP(s) con lp2hp(num, den):

num_hp, den_hp = sig.lp2hp(num_lp, den_lp)

display(Markdown('#####'))
display(Markdown('-- $z_i$: Ceros obtenidos con cheb1ap(N, rp) coincidentes con los obtenidos con roots(p) de forma analítica --'))
display(Markdown('#####'))

display(Math(r'z_{1,2}= ' + sp.latex(np.roots(num_hp))))

display(Markdown('#####'))
display(Markdown('-- $p_i$: Polos en el semiplano izquierdo obtenidos con cheb1ap(N, rp) coincidentes con los obtenidos con roots(p) de forma analítica --'))
display(Markdown('#####'))

display(Math(r'p_{1,2}= ' + sp.latex(np.roots(den_hp))))

display(Markdown('#####'))
display(Markdown('***'))

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos (factorizamos) la T(s) en secciones de segundo orden (SOS) con la función tf2sos_analog()
# this_sos = tf2sos_analog(num_hp, den_hp)

t_T_hp= sig.TransferFunction(num_T_hp, den_T_hp)             # numerador y denominador resolución analítica

t_hp= sig.TransferFunction(num_hp, den_hp)                   # numerador y denominador con cheb1ap()

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T_HP(s) en formato T_HP(s)= num/den con la funcion pretty_print_lti(num, den):

display(Markdown('#####'))
display(Markdown('-- Función transferencia $T_{HP}(s)$ analítica NORMALIZADA: --'))
display(Markdown('#####'))

pretty_print_lti(num_T_hp, den_T_hp)

display(Markdown('#####'))
display(Markdown('-- Función transferencia $T_{HP}(s)$ con cheb1ap(N, rp) NORMALIZADA: --'))
display(Markdown('#####'))

pretty_print_lti(num_hp, den_hp)

display(Markdown('#####'))
display(Markdown('Función transferencia $T_{HP}(s)$ en términos de $k$, $\omega_0$ y $Q$ NORMALIZADA:'))
display(Markdown('#####'))

display(Math(r'$T_{HP}(s)= k \frac{ s^2 \frac{\omega_0}{Q} } { s^2 + s \frac{\omega_0}{Q} + \omega_0^2 } $'))
display(Markdown('#####'))

omega_0= np.sqrt(den_hp[2])
Q= omega_0/den_hp[1]
k= num_hp[0]/omega_0*Q

s = sp.symbols('s', complex=True)

print_latex('T_{HP}(s)= ' + sp.latex(N(k, 4)) + ' \\frac{ ' + sp.latex(s**2) + '\\frac{' + sp.latex(N(omega_0, 4)) + '}{' + sp.latex(N(Q, 4)) + '}' + '}{' + sp.latex(s**2) + ' + ' + sp.latex(s) + '\\frac{' + sp.latex(N(omega_0, 4)) + '}{' + sp.latex(N(Q, 4)) + '}' + ' + ' + sp.latex(N(omega_0**2, 4)) + '}')

display(Markdown('#####'))
display(Markdown('***'))

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la función analyze_sys() que tiene, 
# entre otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

display(Markdown('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --'))
display(Markdown('#####'))

analyze_sys(t_hp, sys_name='High Pass Filter Chevyshev Order 2')

#-------------------------------------------------------------------------------------------------------------------#
```

#####
##### En la primera figura, se observa la respuesta de módulo de $T_{HP}(s)$.
##### Se observa que la pendiente de la banda de detención es $+40\frac{dB}{dec}$ debido al aporte de las singularidades del numerador mientras que en la banda de paso se ve compensada con el aporte de las singularidades del denominador debido a que el par de polos complejos conjugados aporta $−40\frac{dB}{dec}$.
#####
##### En la segunda figura, se observa la respuesta de fase de $T_{HP}(s)$.
##### Se trata de una función sigmoidea monótonamente decreciente cuyo desarrollo de fase total es de $-\pi rad$ producto del aporte de fase del numerador y del denominador: $-\frac{\pi}{2}$ y $+\frac{\pi}{2}$ respectivamente, obteniendo el aporte total como la diferencia entre ellos.
#####
##### En la tercera figura, se observa el diagrama de polos y ceros de $T_{HP}(s)$ que se corresponde con lo mencionado anteriormente en el análisis de singularidades.

##### En la cuarta figura, se observa la respuesta de retardo de grupo de $T_{HP}(s)$.

##### Observamos que en $\omega_0$, el retardo es máximo en concordancia con la pendiente de la recta tangente en ese punto.
##### A medida que nos desplazamos de dicha frecuencia, el retardo disminuye hasta anularse en los puntos en los que la pendiente de la recta tangente es mínima.

```python
#-------------------------------------------------------------------------------------------------------------------#

# Calculamos transferencia desnormalizada: producto de la desnormalización y de dejar mónico el polinomio
# denominador, los términos quedan afectados por los siguientes factores (Ver resolución analítica T_HP(s)
# desnormalizada)

num_T_hp_ = num_T_hp
den_T_hp_ = den_T_hp * [1, Omega_omega, Omega_omega**2]

t_T_hp_= sig.TransferFunction(num_T_hp_, den_T_hp_)

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

display(Markdown('#####'))
display(Markdown('-- Función transferencia T(s) analítica DESNORMALIZADA: --'))
display(Markdown('#####'))

pretty_print_lti(num_T_hp_, den_T_hp_)

display(Markdown('#####'))
display(Markdown('***'))
display(Markdown('#####'))

#-------------------------------------------------------------------------------------------------------------------#

display(Markdown('#####'))
display(Markdown('Función transferencia $T_{HP}(s)$ en términos de $k$, $\omega_0$ y $Q$ NORMALIZADA:'))
display(Markdown('#####'))

display(Math(r'$T_{HP}(s)= k \frac{ s^2 \frac{\omega_0}{Q} } { s^2 + s \frac{\omega_0}{Q} + \omega_0^2 } $'))
display(Markdown('#####'))

omega_0_d= np.sqrt(den_T_hp_[2])
Q_d= omega_0_d/den_T_hp_[1]
k_d= num_T_hp_[0]/omega_0_d*Q_d

s = sp.symbols('s', complex=True)

print_latex('T_{HP}(s)= ' + sp.latex(N(k, 4)) + ' \\frac{ ' + sp.latex(s**2) + '\\frac{' + sp.latex(N(omega_0_d, 4)) + '}{' + sp.latex(N(Q_d, 4)) + '}' + '}{' + sp.latex(s**2) + ' + ' + sp.latex(s) + '\\frac{' + sp.latex(N(omega_0_d, 4)) + '}{' + sp.latex(N(Q_d, 4)) + '}' + ' + ' + sp.latex(N(omega_0_d**2, 4)) + '}')

display(Markdown('#####'))
display(Markdown('***'))

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

display(Markdown('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --'))
display(Markdown('#####'))

analyze_sys(t_T_hp_, sys_name='High Pass Filter Chevyshev Order 2')

#-------------------------------------------------------------------------------------------------------------------#
```

```python
#-----------------------------------------------------------------------------------------------------------------------#

import csv
import matplotlib as mpl
import matplotlib.pyplot as plt

fig_sz_x = 10
fig_sz_y = 9
fig_dpi = 150 # dpi

fig_font_size = 12

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
#plt.rcParams.update({'font.size':fig_font_size})

#-----------------------------------------------------------------------------------------------------------------------#
```

```python
#-----------------------------------------------------------------------------------------------------------------------#

ruta_archivo = 'TPL1_Respuesta_modulo_analizador.csv'

freq_mod= []
resp_mod = []

with open(ruta_archivo, 'r') as archivo_csv:
    lector_csv = csv.reader(archivo_csv)
    
    qty_rows_values = 100
    qty_rows_titles = 3
    qty_rows_blanks = 1
    
    qty_rows_freq_vs_freq = qty_rows_titles + qty_rows_values
    qty_rows_mod_vs_freq = qty_rows_titles + qty_rows_values
    qty_rows_total = qty_rows_freq_vs_freq + qty_rows_blanks + qty_rows_mod_vs_freq
    row_index = 0
    
    for row_freq_resp in lector_csv:
        row_index += 1
        if ( row_index > ( qty_rows_freq_vs_freq + qty_rows_blanks + qty_rows_titles ) and row_index <= qty_rows_total ):
            freq_mod_data, resp_mod_data = row_freq_resp
            freq_mod.append(float(freq_mod_data))
            resp_mod.append(float(resp_mod_data))

figura_1= plt.figure()
plt.plot(freq_mod, resp_mod)

plt.xscale('log')

plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Ganancia [dB]')

dresp= 0.5
plt.xlim(min(freq_mod), max(freq_mod))
plt.ylim(min(resp_mod), max(resp_mod)+dresp)

plt.grid(True, which= 'both')

plt.title('Respuesta en Frecuencia - Filtro Pasa Altos')

plt.show()

#-----------------------------------------------------------------------------------------------------------------------#
```
