### <span style="text-decoration: underline;">Introducción</span>
#####

##### El trabajo práctico implica la implementación de una función transferencia con el circuito integrado [UAF42](https://www.ti.com/lit/ds/symlink/uaf42.pdf?ts=1653513515595&ref_url=https%253A%252F%252Fwww.ti.com%252Fproduct%252FUAF42%253Futm_source%253Dsupplyframe%2526utm_medium%253DSEP%2526utm_campaign%253Dnot_alldatasheet%2526DCM%253Dyes%2526dclid%253DCOTE4LXH-_cCFaCFlQIdQoELbA) que posee una característica especial: capacitores integrados de muy baja tolerancia, aproximadamente del 1%. Esto permite obtener respuestas en frecuencia del filtro sumamente similares a las simulaciones. 

##### A continuación se parte el diseño en 3 secciones de forma tal de tener instancias intermedias para consultar a los docentes sobre cuestiones de implementación y armado.

##### Tomando como referencias las ecuaciones de $|T(jω)|^2$ en su aproximación de Chevyshev [$^{(1)}$](#1) [$^{(2)}$](#2):
#####

\begin{equation}
|T(jΩ)|^2= T(jΩ)*T(-jΩ)= {T(s)*T(-s)}|_{s= jΩ}= \frac{1}{1+ε^2*C_n^2(Ω)}
\end{equation}

#####
##### Podemos entonces establecer la atenuación en función de la ecuación anterior como la inversa de $|T(jΩ)|^2$:
#####

\begin{equation}
|α|^2= \frac {1}{|T(jΩ)|^2}= \frac{1}{\frac{1}{1+ε^2*C_n^2(Ω)}}= 1+ε^2*C_n^2(Ω)
\end{equation}

#####
##### Considerando que $|α|_{dB} \triangleq α_{dB}$, podemos expresar la ecuación en dB [$^{(3)}$](#3):
#####

\begin{equation}
α_{dB}= 10*log[1+ε^2*C_n^2(Ω)]
\end{equation}

#####
##### Despejando $ε^2$ de la ecuación anterior, obtenemos [$^{(4)}$](#4):
#####

\begin{equation}
ε^2= \frac{10^{\frac{α_{máx}}{10}}-1}{C_n^2(Ω)}
\end{equation}

##### Considerando la máxima atenuación requerida por la plantilla $α_{máx}$ en el vértice de fin de la banda de paso $Ω_{p_n}$, obtenemos $ε^2$:
#####

\begin{equation}
ε^2= \frac{10^{\frac{α_{máx}}{10}}-1}{C_n^2(Ω_{p_n})}= \frac{10^{\frac{20}{10}}-1}{1}
\end{equation}

\begin{equation}
\boxed{ε^2= 0,2589}
\end{equation}

\begin{equation}
\boxed{ε= 0,5088}
\end{equation}

##### Considerando la mínima atenuación requerida por la plantilla $α_{mín}$ en el vértice del inicio de la banda de detenida $Ω_{s_n}$, iteramos N de forma tal que la atenuación sea mayor o igual a obtenemos $α_{mín}$:

\end{equation}

#### <span style="text-decoration: underline;">Cálculo de la función transferencia $T_{HP}(s)$</span>
#####

##### Recordando que $|T(jΩ)|^2= T(jΩ)*T(-jΩ)= {T(s)*T(-s)}|_{s= jΩ}= \frac{1}{1+ε^2*C_n^2(Ω)}$ y habiendo obtenido $ε^2$ y N, procedemos a calcular los polinomios de orden N de Chevyshev definidos como:
#####

\begin{equation}
C_n^2= cosh^2(n*cosh^{-1}(Ω_{s}))
\end{equation}

#####
##### Utilizando identidades trigonómetricas [$^{(5)}$](#5), calculamos el polinomio para el grado N obtenido:
#####

\begin{equation}
C_0(Ω)= 1 
\end{equation}

\begin{equation}
C_1(Ω)= Ω 
\end{equation}

\begin{equation}
C_n(Ω)= 2*Ω*C_{n-1}(Ω) - C_{n-2}(Ω) 
\end{equation}

\begin{equation}
C_2(Ω)= 2*Ω*Ω - 1= 2*Ω^2-1 
\end{equation}

#####
##### Reemplazando el polinomio de grado 2 en la ecuación de $|T(jΩ)|^2= \frac{1}{1+ε^2*C_n^2(Ω)}$, operamos algebraicamente:
#####

\begin{equation}
|T(jΩ)|^2_{Ω=\frac{$}{j}}= \frac{1}{1+ε^2*C_n^2(Ω)}= \frac{1}{1+ε^2*C_2^2(Ω)}= \frac{1}{1+ε^2*(2*Ω^2-1)^2}
\end{equation}

\begin{equation}
|T(jΩ)|^2_{Ω=\frac{$}{j}}= \frac{1}{1+ε^2*(2*Ω^2-1)^2}
\end{equation}

\begin{equation}
|T(jΩ)|^2_{Ω=\frac{$}{j}}= \frac{1}{1+ε^2*(2*Ω^2-1)*(2*Ω^2-1)}
\end{equation}

\begin{equation}
|T(jΩ)|^2_{Ω=\frac{$}{j}}= \frac{1}{1+ε^2*(4*Ω^4-2*Ω^2-2*Ω^2+1)}
\end{equation}

\begin{equation}
|T(jΩ)|^2_{Ω=\frac{$}{j}}= \frac{1}{1+ε^2*(4*Ω^4-4*Ω^2+1)}
\end{equation}

#####
##### Reemplazando $Ω$ por $\frac{$}{j}$, obtenemos:
#####

\begin{equation}
T($)*T(-$)= \frac{1}{1+ε^2*\left[4*(\frac{$}{j})^4-4*(\frac{$}{j})^2+1\right]}
\end{equation}
 
\begin{equation}
T($)*T(-$)= \frac{1}{1+ε^2*\left[4*$^4+4*$^2+1\right]}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{1}{1+4*ε^2*^$4+4*ε^2*$^2+ε^2}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{1}{4*ε^2*$^4+4*ε^2*$^2+ε^2+1}
\end{equation}

#####
##### Dejamos mónico el polinomio denominador:
#####

\begin{equation}
T($)*T(-$)= \frac{\frac{1}{4*ε^2}}{\frac{4*ε^2*$^4+4*ε^2*$^2+ε^2+1}{4*ε^2}}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{\frac{1}{4*ε^2}}{$^4+$^2+\frac{ε^2+1}{4*ε^2}}
\end{equation}

#####
##### Definimos los siguientes coeficientes:
#####

\begin{equation}
a= \frac{1}{4*ε^2}
\end{equation}

\begin{equation}
b= \frac{ε^2+1}{4*ε^2}= \frac{ε^2}{4*ε^2} + \frac{1}{4*ε^2}= \frac{1}{4} + \frac{1}{4*ε^2}= \frac{1}{4} + a  
\end{equation}

#####
##### Llegamos a la siguiente expresión en términos de los coeficientes a y b:
#####

\begin{equation}
T($)*T(-$)= \frac{a}{$^4+$^2+b}
\end{equation}

#####
##### Expresamos ahora de forma general en secciones de segundo orden:
#####

\begin{equation}
T($)*T(-$)= \frac{α}{$^2 + β*$ + γ}*\frac{α}{$^2 - β*$ + γ}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{α^2}{$^4 -β*$^3 +γ*$^2 +β*$^3 -β^2*$^2 +β*γ*$ +γ*$^2 -β*γ*$ +γ^2}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{α^2}{$^4 +(-β+β)*$^3 +(2*γ-β^2)*$^2 +(β*γ-β*γ)*$ +γ^2}
\end{equation}

\begin{equation}
T($)*T(-$)= \frac{α^2}{$^4 +(2*γ-β^2)*$^2 +γ^2}
\end{equation}

######
##### Igualamos los coeficientes de ambas expresiones y resolvemos el sistema de ecuaciones:
#####

\begin{equation}
α^2= a= \frac{1}{4*ε^2}
\end{equation}

\begin{equation}
α= \sqrt{\frac{1}{4*ε^2}}= \frac{1}{2*ε}= \frac{1}{2*0,5088}= 0.9827
\end{equation}

\begin{equation}
2*γ-β^2= 1
\end{equation}

\begin{equation}
β= \sqrt{2*γ-1}
\end{equation}

\begin{equation}
γ^2= b
\end{equation}

\begin{equation}
γ^2= \frac{1}{4} + a  
\end{equation}

\begin{equation}
γ= \sqrt{\frac{1}{4} + a}= \sqrt{\frac{1}{4} + α^2}= \sqrt{\frac{1}{4} + 0.9827^2}= 1,1026 
\end{equation}

\begin{equation}
β= \sqrt{2*γ-1}= \sqrt{2*1,1026-1}= 1,0978
\end{equation}

#####
##### Podemos expresar entonces la transferencia pasa bajos prototipo normalizada $T_{LP}(\$)$:
#####

\begin{equation}
\boxed{T_{LP}($)=\frac{\frac{1}{2*ε}}{$^2 + \sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*$ + \frac{\sqrt{ε^2+1}}{2*ε}}= \frac{0.9827}{$^2 + 1,0978*$ + 1,1026}}
\end{equation}

#####
##### Aplicamos el núcleo de la transformación $K(s)= \$= \frac{1}{s}$ para obtener la transferencia pasa altos objetivo normalizada $T_{HP}(s)$:
#####

\begin{equation}
T(s)= \frac {\frac{1}{2*ε}}{(\frac{1}{s})^2 + \sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*(\frac{1}{s}) + \frac{\sqrt{ε^2+1}}{2*ε}} 
\end{equation}

\begin{equation}
T(s)= \frac{\frac{1}{2*ε}}{\frac{1}{s^2} + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{s} + \frac{\sqrt{ε^2+1}}{2*ε}}
\end{equation}

\begin{equation}
T(s)= \frac{\frac{1}{2*ε}}{\frac{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}{s^2}}= \frac{\frac{1}{2*ε}*s^2}{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}
\end{equation}

\begin{equation}
T(s)= \frac{\frac{\frac{1}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2}{\frac{1 +\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}*s +\frac{\sqrt{ε^2+1}}{2*ε}*s^2}{\frac{\sqrt{ε^2+1}}{2*ε}}}
\end{equation}

\begin{equation}
T(s)= \frac {\frac{\frac{1}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2} {\frac{1}{\frac{\sqrt{ε^2+1}}{2*ε}} +\frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s +\frac{\frac{\sqrt{ε^2+1}}{2*ε}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s^2}
\end{equation}

\begin{equation}
T(s)= \frac {\frac{1}{\sqrt{ε^2+1}}*s^2} {\frac{2*ε}{\sqrt{ε^2+1}} +\frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s +s^2}
\end{equation}

\begin{equation}
T_{HP}(s)=\frac{\frac 1{\sqrt{ε^2+1}}*s^2}{s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*s + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

\begin{equation}
\boxed{T_{HP}(s)= \frac {0,8913*s^2} {s^2 +0,9957*s +0,9069}}
\end{equation}

#####
##### Considerando $T_{HP}(s)$ en términos de $\omega_0$ y $Q$, podemos expresarla de forma general de la siguiente manera:
#####

\begin{equation}
T_{HP}(s)=\frac{k * s^2} {s^2 + \frac{ω_0}{Q}*s + ω_0^2}
\end{equation}


#####
##### Obtenemos los valores de $\omega_0$ y $Q$ para este caso particular:
#####

\begin{equation}
ω_0^2= 0,9069
\end{equation}

\begin{equation}
ω_0= 0,9523
\end{equation}

\begin{equation}
\frac{ω_0}{Q}= 0,9957
\end{equation}

\begin{equation}
Q= \frac{ω_0}{0,9957}= \frac{0,9523}{0,9957}= 0,9564
\end{equation}

#####
##### Obtenemos $T_{HP}(s)$ en términos de $\omega_0$ y $Q$, para este caso particular:
#####

\begin{equation}
\boxed{T_{HP}(s)=\frac{0,8913*s^2} {s^2 + \frac{0,9523}{0,9564}*s + 0,9069}}
\end{equation}

#####
##### La $T_{HP}(s)$ anterior se trata de una transferencia normalizada y por lo general suele omitirse el correspondiente subíndice: lo correcto sería expresar dicha transferencia en términos de $s_N$.
##### Para desnormalizar la $T_{HP}(s)$ recordamos que $s_N= \frac{s}{\Omega_\omega}$ y siendo $\Omega_\omega= 2*\pi*4600 Hz$, obtenemos la transferencia del filtro pasa altos desnormalizada $T_{HP}(s)$:
#####

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * (\frac{s}{\Omega_\omega})^2 }{(\frac{s}{\Omega_\omega})^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*\frac{s}{\Omega_\omega} + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * \frac{1}{\Omega_\omega^2} * s^2 } { \frac{1}{\Omega_\omega^2} * s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}} * \frac{1}{\Omega_\omega} * s + \frac{2*ε}{\sqrt{ε^2+1}}}
\end{equation}

#####
##### Dejamos mónico el polinomio denominador:
#####

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * \frac{\Omega_\omega^2}{\Omega_\omega^2} * s^2} { \frac{\Omega_\omega^2}{\Omega_\omega^2}*s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}}*\frac{\Omega_\omega^2}{\Omega_\omega}*s + \frac{2*ε}{\sqrt{ε^2+1}}*\Omega_\omega^2 }
\end{equation}

\begin{equation}
T_{HP}(s)= \frac{ \frac{1}{\sqrt{ε^2+1}} * s^2} { s^2 + \frac{\sqrt{\frac{\sqrt{ε^2+1}}{ε}-1}}{\frac{\sqrt{ε^2+1}}{2*ε}} * \Omega_\omega * s + \frac{2*ε}{\sqrt{ε^2+1}}*\Omega_\omega^2 }
\end{equation}

\begin{equation}
\boxed{T_{HP}(s)= \frac{0,8913*s^2} {s^2 +2,8777*10^4*s +7,5759*10^8}}
\end{equation}

\end{equation}

\begin{equation}
C_{1n}= C_1*Ω_ω*Ω_z= C_1*ω_p*R= 1000pF * 2*π*4600\frac{rad}{s} * 50KΩ= 1,4451
\end{equation}

\begin{equation}
C_{2n}= C_2*Ω_ω*Ω_z= C_2*ω_p*R= 1000pF * 2*π*4600\frac{rad}{s} * 50KΩ= 1,4451
\end{equation}

#####

##### Obtenemos los valores de $R_{F1n}$ y $R_{F2n}$ reemplazando los valores normalizados de los componentes en la ecuación de $ω_0^2$:

#####

\begin{equation}
ω_0^2= \frac{R_{2n}}{R_{1n}*R_{F1n}*R_{F2n}*C_{1n}*C_{2n}} \frac{1}{4*10^{-5}*R_{F1}*R_{F2}*C_{1n}*C_{2n}} 
\end{equation}

\begin{equation}
R_{F1n}*R_{F2n}= \frac{R_{2n}}{R_{1n}*ω_0^2*C_{1n}*C_{2n}} 
\end{equation}

\begin{equation}
R_{F1n}*R_{F2n}= \frac{1}{1*0,9069*1,4451*1,4451} 
\end{equation}

#####

##### Considerando $R_{F1n}= R_{F2n}$:

#####

\begin{equation}
R_{F1n}= R_{F2n}= \sqrt{\frac{1}{0,9069*1,4451*1,4451}}= 0,7266
\end{equation}

#####

##### Obtenemos el valor de $R_{Gn}$ considerando ${A_{HP}}\Bigr|_{ω \rightarrow \infty}  \rightarrow 0,8913$ y reemplazando los valores normalizados de los componentes en la ecuación de ${A_{HP}}$:

#####

\begin{equation}
A_{HP}= \frac{R_{2n}}{R_{Gn}}
\end{equation}

\begin{equation}
R_{Gn}= \frac{R_{2n}}{A_{HP}} 
\end{equation}

\begin{equation}
R_{Gn}= \frac{1}{0,8913}= 1,1219 
\end{equation}

##### 

##### Obtenemos el valor de $R_{Qn}$ reemplazando los valores normalizados de los componentes en la ecuación de $Q$:

#####

\begin{equation}
Q= \left( 1 + \frac{R_{4n}}{R_{Qn}} \right) * \frac{1}{\left(\frac{1}{R_{1n}} + \frac{1}{R_{2n}} + \frac{1}{R_{Gn}} \right)} * \sqrt{\frac{R_{F1n}*C_{1n}}{R_{1n}*R_{2n}*R_{F2n}*C_{2n}}}
\end{equation}

\begin{equation}
\frac{Q}{\frac{1}{\left(\frac{1}{R_{1n}} + \frac{1}{R_{2n}} + \frac{1}{R_{Gn}} \right)} * \sqrt{\frac{R_{F1n}*C_{1n}}{R_{1n}*R_{2n}*R_{F2n}*C_{2n}}}}= 1 + \frac{R_{4n}}{R_{Qn}}
\end{equation}

\begin{equation}
\frac{Q}{\frac{1}{\left(\frac{1}{R_{1n}} + \frac{1}{R_{2n}} + \frac{1}{R_{Gn}} \right)} * \sqrt{\frac{R_{F1n}*C_{1n}}{R_{1n}*R_{2n}*R_{F2n}*C_{2n}}}}-1= \frac{R_{4n}}{R_{Qn}}
\end{equation}

\begin{equation}
R_{Qn}= \frac{R_{4n}}{\frac{Q}{\frac{1}{\left(\frac{1}{R_{1n}} + \frac{1}{R_{2n}} + \frac{1}{R_{Gn}} \right)} * \sqrt{\frac{R_{F1n}*C_{1n}}{R_{1n}*R_{2n}*R_{F2n}*C_{2n}}}}-1}
\end{equation}

\begin{equation}
R_{Qn}= \frac{1}{\frac{0,9564}{\frac{1}{\left(\frac{1}{1} + \frac{1}{1} + \frac{1}{1,1219} \right)} * \sqrt{\frac{0,7266*1,4451}{1*1*0,7266*1,4451}}}-1}= \frac{1} { 0,9564*\left(2 + \frac{1}{1,1219} \right)-1 }
\end{equation}

\begin{equation}
R_{Qn}= 0,5665
\end{equation}

```python
#-------------------------------------------------------------------------------------------------------------------#

# Inicializamos e importamos módulos

# Módulos externos
import matplotlib as mpl
import matplotlib.pyplot as plt
import math as m
import numpy as np
import scipy.signal as sig

fig_sz_x = 10
fig_sz_y = 9
fig_dpi = 150 # dpi

fig_font_size = 12

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
#plt.rcParams.update({'font.size':fig_font_size})

#-------------------------------------------------------------------------------------------------------------------#

# Importamos las funciones de PyTC2

from pytc2.sistemas_lineales import analyze_sys
from pytc2.sistemas_lineales import pretty_print_lti

#-------------------------------------------------------------------------------------------------------------------#

# Datos de plantilla

alfa_max = 1                        # dB
alfa_min = 20                       # dB

f_s_plantilla = 1.2e3               # Hz
f_p_plantilla = 4.6e3               # Hz

omega_p_plantilla = 2*m.pi*f_p_plantilla
norm_w = omega_p_plantilla
omega_p_plantilla /= norm_w
omega_p_pb_prototipo = 1/omega_p_plantilla

omega_s_plantilla = 2*m.pi*f_s_plantilla
omega_s_plantilla /= norm_w
omega_s_pb_prototipo = 1/omega_s_plantilla

print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- Datos de plantilla - Filtro pasa altos objetivo --')
print(' ')

print('α_min = {}'.format(alfa_min))
print('α_max = {}'.format(alfa_max))
print('fs = {}'.format(f_s_plantilla))
print('fp = {}'.format(f_p_plantilla))

print(' ')

print('ωp_n = {:3.4f}'.format(omega_p_plantilla))
print('ωs_n = {:3.4f}'.format(omega_s_plantilla))

print(' ')
print('-- Dominio transformado - Filtro pasa bajos prototipo --')
print(' ')

print('ωp = {:3.4f}'.format(omega_p_pb_prototipo))
print('ωs = {:3.4f}'.format(omega_s_pb_prototipo))

#-------------------------------------------------------------------------------------------------------------------#

# Calculamos e^2

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')
print('-- ε: Grado de libertad de la funcion --')
print(' ')

print('ε = sqrt(10^(α_max/10)-1)')
eps = np.sqrt(10**(alfa_max/10)-1)

print(' ')

print('ε = {:3.4f}'.format(eps))
print('ε^2 = {:3.4f}'.format(eps**2))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Calculamos N iterando alfa_min

print(' ')
print('-- N: Orden del filtro --')
print(' ')
print('α_min = 10*log(1 + ε^2*C_n^2) con C_n^2= cosh^2(N*cosh^(-1)(ω_s))')
print(' ')

for N in range(1,10):
    C_nn = (np.cosh(N*(np.arccosh(omega_s_pb_prototipo))))**2
    att = 10*np.log10(1+(eps**2)*C_nn)    
    print('Para N= {:d}: ¿att = {:3.4f} dB >= α_min= {:d}?'.format(N, att, alfa_min))
    if att >= alfa_min:
        break

print(' ')

# Elegimos N de forma tal que α_min >= 20 dB:
print('N= {:d}'.format(N))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# De la resolución analítica, obtenemos |T(jw)|^2 |w=s/j = 1 / (1 + ε^2*C_n^2) = T(s)*T(-s) 

#Cn2 = 2w^2 - 1                 # Polinomio grado 2 de Chevyshev
Cn2 = [2, 0, -1]                # Coeficientes del polinomio grado 2 de Chevyshev
Cn4 = np.polymul(Cn2, Cn2)      # Elevamos al cuadrado el denominador
Cn4 = Cn4 * eps**2              # Multiplicamos por ε^2
Cn4[4]= Cn4[4] + 1              # Sumamos 1 al término independiente
Cn4[2]= -Cn4[2]                 # Cambiamos de signo al coeficiente del término cuadrático producto del reeemplazo de w^2= s^2/j^2
Cn4 = Cn4 / (4*eps**2)          # Hacemos mónico el polinomio

p_T_lp= np.roots(Cn4)

den_T_lp= np.poly([p_T_lp for p_T_lp in p_T_lp if p_T_lp.real < 0])
num_T_lp= [m.sqrt(1/(4*eps**2))]
#print(num_T_lp)
#print(den_T_lp)

num_T_hp, den_T_hp= sig.lp2hp(num_T_lp, den_T_lp)
#print(num_T_hp)
#print(den_T_hp)

print(' ')
print('-- z_i: Ceros obtenidos con roots(p) del polinomio numerador --')
print('-- obtenido de forma analítica --')
print(' ')

print('z1,2= {}'.format(np.roots(num_T_hp)))

print(' ')
print('-- p_i: Polos en el semiplano izquierdo obtenidos con roots(p) --')
print('-- del polinomio denominador obtenido de forma analítica --')
print(' ')

print('p1,2= {}'.format(np.roots(den_T_hp)))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Obtenemos singularidades con la función de aproximación de Chevyshev del 
# filtro pasa bajos prototipo: cheb1ap(N, rp)

z, p, k = sig.cheb1ap(N, alfa_max)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos las singularidades que nos devuelve cheb1ap(N, rp) para obtener 
# el numerador y el denominador de la T_LP(s) del filtro pasa bajos prototipo 
# con zpk2tf(z,p,k):

num_lp, den_lp = sig.zpk2tf(z,p,k)

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos la T_LP(s) del filtro pasa bajos prototipo para obtener 
# el numerador y el denominador de la T_HP(s) con lp2hp(num, den):

num_hp, den_hp = sig.lp2hp(num_lp, den_lp)

print(' ')
print('-- z_i: Ceros obtenidos con cheb1ap(N, rp) coincidentes --')
print('con los obtenidos con roots(p) de forma analítica --')
print(' ')

print('z1,2= {}'.format(np.roots(num_hp)))

print(' ')
print('-- p_i: Polos en el semiplano izquierdo obtenidos con cheb1ap(N, rp) --')
print('-- coincidentes con los obtenidos con roots(p) de forma analítica --')
print(' ')

print('p1,2= {}'.format(np.roots(den_hp)))

print(' ')
print('#---------------------------------------------------------------------------------------------#')

#-------------------------------------------------------------------------------------------------------------------#

# Transformamos (factorizamos) la T(s) en secciones de segundo orden (SOS) con la función tf2sos_analog()
# this_sos = tf2sos_analog(num_hp, den_hp)

t_T_hp= sig.TransferFunction(num_T_hp, den_T_hp)             # numerador y denominador resolución analítica

t_hp= sig.TransferFunction(num_hp, den_hp)                   # numerador y denominador con cheb1ap()

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('-- Función transferencia T(s) analítica NORMALIZADA: --')
print(' ')

pretty_print_lti(num_T_hp, den_T_hp)

print(' ')
print('-- Función transferencia T(s) con cheb1ap(N, rp) NORMALIZADA: --')
print(' ')

pretty_print_lti(num_hp, den_hp)

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

print('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --')
print(' ')

analyze_sys(t_hp, sys_name='Filtro pasa alto Chevyshev 2do orden')

#-------------------------------------------------------------------------------------------------------------------#
```

```python
#-------------------------------------------------------------------------------------------------------------------#

# Calculamos transferencia desnormalizada: producto de la desnormalización y de dejar mónico el polinomio
# denominador, los términos quedan afectados por los siguientes factores (Ver resolución analítica T_HP(s)
# desnormalizada)

num_T_hp_ = num_T_hp
den_T_hp_ = den_T_hp * [1, norm_w, norm_w**2]

t_T_hp_= sig.TransferFunction(num_T_hp_, den_T_hp_)

#-------------------------------------------------------------------------------------------------------------------#

# Imprimimos la T(s) en formato T(s)= num/den con la funcion pretty_print_lti(num, den):

print(' ')
print('-- Función transferencia T(s) analítica DESNORMALIZADA: --')
print(' ')

pretty_print_lti(num_T_hp_, den_T_hp_)

print(' ')
print('#---------------------------------------------------------------------------------------------#')
print(' ')

#-------------------------------------------------------------------------------------------------------------------#

# Representamos módulo, fase, diagrama de polos y ceros y retardo de fase con la analyze_sys() que tiene, entre
# otras cosas, indicadores (w0 y Q) de las singularidades en el diagrama de polos y ceros

print('-- Módulo, fase, diagrama de polos y ceros y retardo de fase --')
print(' ')

analyze_sys(t_T_hp_, sys_name='Filtro pasa alto Chevyshev 2do orden')

#-------------------------------------------------------------------------------------------------------------------#
```
