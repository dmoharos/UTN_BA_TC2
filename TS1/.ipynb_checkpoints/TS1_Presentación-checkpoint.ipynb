\begin{equation}
I_1= \frac{V_1-V_A}{Z_1}= \frac{V_A-V_2}{Z_2}
\end{equation}

\begin{equation}
(V_1-V_A)*Z_2= (V_A-V_2)*Z_1
\end{equation}

\begin{equation}
V_1*Z_2-V_A*Z_2= V_A*Z_1-V_2*Z_1
\end{equation}

\begin{equation}
V_1*Z_2+V_2*Z_1= V_A*(Z_1+Z_2)
\end{equation}

\begin{equation}
V_A= \frac{V_1*Z_2+V_2*Z_1}{Z_1+Z_2}
\end{equation}

\begin{equation}
[V_A]= \frac{V*Ω+V*Ω}{Ω}= \frac{V*Ω}{Ω}= [V] 
\end{equation}

\begin{equation}
VZ_3= V_1\frac{Z_3}{Z_3+Z_4}= V_A= \frac{V_1*Z_2+V_2*Z_1}{Z_1+Z_2}
\end{equation}

\begin{equation}
(V_1*Z_2+V_2*Z_1)*(Z_3+Z_4)= V_1*Z_3*(Z_1+Z_2)
\end{equation}

\begin{equation}
V_1*Z_2*Z_3+V_1*Z_2*Z_4+V_2*Z_1*Z_3+V_2*Z_1*Z_4= V_1*Z_1*Z_3+V_1*Z_2*Z_3
\end{equation}
