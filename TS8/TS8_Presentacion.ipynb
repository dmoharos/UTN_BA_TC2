###
##### La transferencia del filtro analógico pasa bajos de segundo orden está dada por:
###

\begin{equation*}
H_{LP}(s)= \frac{{\omega_0}^2}{s^2 +s\frac{\omega_0}{Q} +{\omega_0}^2}
\end{equation*}

###
##### Para el caso particular de la transferencia Butterworth, consideramos: 
###
##### Norma de pulsación: $\Omega_\omega = \omega_0$
###
##### Apertura angular $\measuredangle_{\sigma, \omega_0}$ de las singularidades del denominador: $\psi= \frac{\pi}{4}$
###
##### Factor de selectividad: $Q= \frac{\sqrt{2}}{2}$
###
##### Obtenemos finalmente:
###

\begin{equation*}
H(s)= \frac{1}{s^2 +s\frac{1}{\frac{\sqrt{2}}{2}} +1}
\end{equation*}

###

\begin{equation*}
H(s)= \frac{1}{s^2 +s\sqrt{2} +1}
\end{equation*}

###
##### La transferencia del filtro digital está dada por:
###

\begin{equation*}
H(z)= H(s)|_{s= k\frac{ z -1 }{ z +1 }= 2f_s\frac{ z -1 }{ z +1 }}
\end{equation*}

###
##### Reemplazamos el núcleo de la transformación en $H(s)$: 
###

\begin{equation*}
H(z)= \frac{1}{\left( k\frac{z-1}{z+1}\right)^2 +\left( k\frac{ z -1 }{ z +1 } \right)\sqrt{2} +1}
\end{equation*}

###
##### Aplicamos el cuadrado a cada término:
###

\begin{equation*}
H(z)= \frac{1}{ k^2\frac{ (z-1)^2 }{ (z+1)^2 } +k\frac{ z -1 }{ z +1 }\sqrt{2} +1}
\end{equation*}

###
##### Obtenemos el denominador común:
###

\begin{equation*}
H(z)= \frac{1}{ \frac{k^2(z-1)^2 +k(z-1)*(z+1)\sqrt{2} +(z+1)^2 }{ (z+1)^2 } }
\end{equation*}

###

\begin{equation*}
H(z)= \frac{ (z+1)^2 }{ k^2(z-1)^2 +k(z-1)(z+1)\sqrt{2} +(z+1)^2 }
\end{equation*}

###
##### Desarrollamos los términos:
###

\begin{equation*}
H(z)= \frac{ (z^2 +2z +1) } { k^2(z^2 -2z +1) +k(z^2 +z -z -1)\sqrt{2} +(z^2 +2z +1) }
\end{equation*}

###
##### Agrupamos coeficientes:
###

\begin{equation*}
H(z)= \frac{ ( z^2 +2z +1 ) } { z^2( k^2 +k\sqrt{2} +1 ) + z( 2 -2k^2 ) + ( k^2 -k\sqrt{2} +1 ) }
\end{equation*}

###
##### Dejamos mónico el polinomio denominador:
###

\begin{equation*}
H(z)=  \frac{1}{ k^2 +k\sqrt{2} +1 } * \frac{ z^2 +2z +1 } { z^2 + z\left( \frac{ 2 -2k^2 }{ k^2 +k\sqrt{2} +1 } \right) +\left( \frac{ k^2 -k\sqrt{2} +1 }{ k^2 +k\sqrt{2} +1 } \right)}
\end{equation*}

###
##### Reemplazamos $k$ por $2fs$:
###

\begin{equation*}
H(z)=  \frac{1}{ 4fs^2 +2\sqrt{2}fs +1 } \frac{ z^2 +2z +1 } { z^2 + z\left( \frac{ 2 -8fs^2 }{ 4fs^2 +2\sqrt{2}fs +1 }\right) +\left( \frac{ 4fs^2 -2\sqrt{2}fs +1 }{ 4fs^2 +2\sqrt{2}fs +1 }\right)}
\end{equation*}

###
##### Para $fs= 100kHz$ obtenemos $H(z)$:
###

\begin{equation*}
H(z)|_{fs= 100kHz}=  \frac{1}{ 4*(100kHz)^2 +2\sqrt{2}*100kHz +1 } * \frac{ (z^2 +2z +1) } { z^2 + z\left[ \frac{ 2 -8*(100kHz)^2 }{ 4*(100kHz)^2 +2\sqrt{2}*100kHz +1 }\right] +\left[\frac{ 4*(100kHz)^2 -2\sqrt{2}*100kHz +1 }{ 4*(100kHz)^2 +2\sqrt{2}*100kHz +1 }\right]}
\end{equation*}

###

\begin{equation*}
H(z)|_{fs= 100kHz}=  2,5*10^{-11} * \frac{ z^2 +2z +1 } { z^2 -2z + 1 }
\end{equation*}

###

```python
# ---------------------------------------------------------------------------------------- #
# Librerias utilizadas
# ---------------------------------------------------------------------------------------- #

import sympy as sp
import numpy as np
import scipy.signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
from pytc2.sistemas_lineales import group_delay, analyze_sys, pzmap
from pytc2.general import print_latex

# ----------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# Ajustes de visualización
# ---------------------------------------------------------------------------------------- #

fig_sz_x = 10
fig_sz_y = 9
fig_dpi = 150 # dpi

fig_font_size = 12

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
#plt.rcParams.update({'font.size':fig_font_size})

# ---------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# Resolución simbólica
# ---------------------------------------------------------------------------------------- #

s, z = sp.symbols('s z', complex=True)
k, fs, Q = sp.symbols('k fs Q', real=True, positive=True)

Ts = 1/(s**2 +s*np.sqrt(2) +1)
fz = k * ( z -1 ) / ( z +1 )
Tz = sp.collect(sp.simplify(sp.expand(Ts.subs(s, fz))), z)

display(Markdown('##'))
display(Markdown('-- Función transferencia del filtro analógico pasa bajos Butterworth de segundo orden: --'))
display(Markdown('#####'))
print_latex( 'H_{LP}(s) \Big|_{Butter-2}= ' + sp.latex(Ts) )

display(Markdown('##'))
display(Markdown('-- Función transferencia del filtro digital producto de la transformación bilineal en función de k: --'))
display(Markdown('#####'))
print_latex( 'H_{LP}(z) = ' + sp.latex(Tz) )

display(Markdown('##'))
display(Markdown('-- Función transferencia del filtro digital producto de la transformación bilineal en función de fs: --'))
display(Markdown('#####'))
print_latex( 'H_{LP}(z) = ' + sp.latex(Tz.subs(k, 2*fs)) )
display(Markdown('##'))
# ---------------------------------------------------------------------------------------- #
```

###
##### Las transferencias obtenidas simbólicamente coinciden con las obtenidas analíticamente: existen diferencias en términos de la visualización debido a las aproximaciones que realizan las funciones del paquete SymPy utilizadas.
###

```python
# ---------------------------------------------------------------------------------------- #
# Resolución numérica
# ---------------------------------------------------------------------------------------- #

#-- LPF Butterworth 2º orden --
num_butter_ord_2 = [ 1 ]
den_butter_ord_2 = [ 1, np.sqrt(2), 1 ]
my_tf_butter_ord_2 = sig.TransferFunction( num_butter_ord_2, den_butter_ord_2 )

#-- Frecuencia de corte --
fc = 1e3

#-- Norma de frecuencia --
Omega_omega = fc

#-- Frecuencia de muestreo --
fs = 100e3

#-- Normalización --
fs = fs / Omega_omega

#-- Kernel de la transformación bilineal --
k = 2*fs
            
#-- Coeficientes de denominador de T_LPF(z) --
a2 = k**2 +k*np.sqrt(2) +1
a1 = 2 -2*k**2
a0 = k**2  -k*np.sqrt(2) +1

#-- Generación de la T_LPF(z) --
numz = [1, 2, 1]
denz =  [a2, a1, a0]
my_df = sig.TransferFunction( numz, denz, dt = 1/fs )

#-- Respuestas de módulo, fase, retardo y pzmap
analyze_sys(my_tf_butter_ord_2, "LPF Butterworth 2º orden" )
analyze_sys(my_df, "Filtro Digital con fs = 100 kHz", fs = fs )

# ---------------------------------------------------------------------------------------- #
```

###
##### El prewarping es una técnica utilizada en el diseño de filtros digitales para corregir la distorsión que se puede generar al mapear una frecuencia del dominio analógico al dominio digital mediante la transformación bilineal.
###
##### Debido a que la transformación no es lineal, sino que está dada por la función tangente, los efectos de la transformación se ponen de manifiesto a altas frecuencias producto de la compresión de las mismas en el dominio digital.
###
##### Por lo tanto en la medida que la frecuencia de corte del filtro analógico se acerca a Nyquist, puede requerirse la aplicación de prewarping para corregir dicha distorsión.
###
##### Por lo expuesto, podría ser necesario aplicar prewarping en el caso del LPF analógico muestreado a $fs = 10kHz$. No así en el caso del HPF analógico, muestreado a la misma frecuencia, dado que no cumple con el teorema de Nyquist y por lo tanto las distorsiones que se generen no podrían ser corregidas mediante esta técnica.

\end{equation*}

###
##### Por caso, para $H_1(\Omega)$:
###

\begin{equation*}
\Delta H_1= \frac{ |H( \Omega )| - |H_1( \Omega )| } { |H( \Omega )| }
\end{equation*}

###
##### Reemplazamos los valores correspondientes:
###

\begin{equation*}
0,05 = \frac{ \Omega - 2*sen(\frac{ \Omega }{ 2 }) } { \Omega }
\end{equation*}

###
##### Obtenemos la ecuación a resolver:
###

\begin{equation*}
sen( \frac{ \Omega }{ 2 } ) -0,475 * \Omega = 0
\end{equation*}

###
##### De igual forma, para $H_2(\Omega)$:
###

\begin{equation*}
\Delta H_2= \frac{ |H( \Omega )| - |H_2( \Omega )| } { |H( \Omega )| }
\end{equation*}

###
##### Reemplazamos los valores correspondientes:
###

\begin{equation*}
0,05 = \frac{ \Omega - 2*sen( \Omega ) } { \Omega }
\end{equation*}

###
##### Obtenemos la ecuación a resolver:
###

\begin{equation*}
sen( \Omega ) -0,475 * \Omega = 0
\end{equation*}

###


```python
# ---------------------------------------------------------------------------------------- #
# -- Resolución numérica --
# ---------------------------------------------------------------------------------------- #

#-- Generación de la H(z) a partir de los coeffs del numerador y denominador --

#-- Derivador de 1º orden --
num_1 = np.array( [1, -1] )
den_1 = np.array( [1, 0] )

wrad1, hh1 = sig.freqz( num_1, den_1, worN=1000 )
abs_value1 = abs( hh1 )
ww1 = wrad1 / np.pi

#-- Derivador de 2º orden --
num_2 = np.array( [1, 0, -1] )
den_2 = np.array( [1, 0, 0] )

wrad2, hh2 = sig.freqz( num_2, den_2, worN=1000 )
abs_value2 = abs( hh2 )
ww2 = wrad2 / np.pi

#-- Derivador ideal --
ww3 = np.linspace(0, 1, 1000)
hh3 = ww3 * np.pi

# ---------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# -- Resolución simbólica --
# ---------------------------------------------------------------------------------------- #

Omega = sp.symbols('\Omega')

Omega_h13 = sp.nsolve( sp.sin( Omega / 2 ) -0.475*Omega, Omega, 1 ) / np.pi
Omega_h23 = sp.nsolve( sp.sin( Omega ) -0.475*Omega, Omega, 1 ) / np.pi

hh_31 = Omega_h13 * np.pi
hh_32 = Omega_h23 * np.pi

hh_13 = 2 * np.sin( float(Omega_h13) * np.pi/ 2 )
hh_23 = 2 * np.sin( float(Omega_h23) * np.pi )

# ---------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# -- Respuesta de modulo --
# ---------------------------------------------------------------------------------------- #

plt.close( 'all' )
plt.figure(1)

plt.plot( ww1, abs_value1, label = '$|H_1(\Omega)|_{derivador-orden-1}$' )
plt.plot( ww2, abs_value2, label = '$|H_2(\Omega)|_{derivador-orden-2}$' )
plt.plot( ww3, hh3, label = '$|H_3(\Omega)|_{derivador-ideal}$' )

plt.plot( Omega_h13, hh_31, 'x', color='red', ms=5 )
plt.plot( Omega_h13, hh_13, 'x', color='red', ms=5 )

plt.plot( Omega_h23, hh_32, 'x', color='red', ms=5 )
plt.plot( Omega_h23, hh_23, 'x', color='red', ms=5 )

plt.title( 'Magnitude response' )
plt.xlabel( 'Normalized angular frequency []' )
plt.ylabel( 'Magnitude []' )
plt.grid( which = 'both', axis = 'both' )
plt.legend()
plt.show()

# ---------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# -- Cálculo de diferencia entre valores reales e ideales --
# ---------------------------------------------------------------------------------------- #
error_1 = []
error_2 = []
i = 0
for hh in hh3:
    if hh != 0:
        error_1.append( ( hh - abs_value1[i] ) / hh * 100 )
        error_2.append( ( hh - abs_value2[i] ) / hh * 100 )
        i+= 1
    else:
        error_1.append(0)
        error_2.append(0)
        i+= 1

ww_error_1 = 0
ww_error_2 = 0
hh_error_1 = 0
hh_error_2 = 0 
j = 0
for ww in ww1:
    if error_1[j] <= 5:
        ww_error_1 = ww
        hh_error_1 = error_1[j]
    if error_2[j] <= 5:
        ww_error_2 = ww
        hh_error_2= error_2[j]
    j+= 1

# ---------------------------------------------------------------------------------------- #
```

```python
# ---------------------------------------------------------------------------------------- #
# -- Respuesta de error --
# ---------------------------------------------------------------------------------------- #

plt.figure(2)
plt.plot( ww1, error_1, label = 'Error derivador 1° orden' )
plt.plot( ww2, error_2, label = 'Error derivador 2° orden' )
    
plt.plot( ww_error_1, hh_error_1, 'x', color='red', ms=5 )
plt.plot( ww_error_2, hh_error_2, 'x', color='red', ms=5 )

plt.title( 'Error response' )
plt.xlabel( 'Normalized angular frequency []' )
plt.ylabel( 'Error [%]' )
plt.grid( which = 'both', axis = 'both' )
plt.legend()
plt.show()

# ---------------------------------------------------------------------------------------- #
```

###
##### Obtenemos el esquema a partir de la suma de la señal de entrada con la señal de salida del filtro pasa todo:
###

<img src="TS8_Notch_Filter_Diagrama.png" style="margin-left:auto; margin-right:auto"/>

###
##### Afectamos la transferencia obtenida por un factor de $\frac{1}{2}$ para que no presente ganancia en continua:
###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left[ 1 + H_{All-Pass-Filter}(Z) \right]
\end{equation*}

###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left( 1 + \frac{ \alpha*Z^M +1 }{ Z^M +\alpha } \right)
\end{equation*}

###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left( \frac{ Z^M +\alpha }{ Z^M +\alpha } + \frac{ \alpha*Z^M +1 }{ Z^M +\alpha } \right)
\end{equation*}

###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left( \frac{ Z^M +\alpha +\alpha*Z^M +1 }{ Z^M +\alpha } \right)
\end{equation*}

###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left[ \frac{ Z^M * \left(\alpha +1 \right) +\alpha +1 }{ Z^M +\alpha } \right]
\end{equation*}

###
##### Reemplazando el valor de M, obtenemos el filtro a calcular:
###

\begin{equation*}
H_{Notch}(Z) = \frac{1}{2} * \left[ \frac{ Z^4 * \left(\alpha +1 \right) +\alpha +1 }{ Z^4 +\alpha } \right]
\end{equation*}

###
##### Para distintos valores de $\alpha$ dados, obtendremos distintas respuestas del mismo filtro:
###

```python
# ---------------------------------------------------------------------------------------- #

#-- Frecuencia de muestreo normalizada por f_Nyquist
fs = 2
#-- Vector de valores de alpha arbitrarios cuyos límites se
#-- corresponden con el alpha del inciso anterior y alpha del
#-- all pass filter respectivamente:
alpha = [0.8, 0.825, 0.85, 0.875, 0.9, 0.925, 0.95, 0.975, 1] 

for index, this_alpha in enumerate(alpha):
    
    num = 0.5 * np.array( [ 1+this_alpha, 0, 0, 0, 1+this_alpha ] )
    den = np.array( [ 1, 0, 0, 0, this_alpha ] )

    wrad1, hh1 = sig.freqz( num, den, worN=1000 )    
    abs_value1 = 20 * np.log10( abs( hh1 ) )
    ww1 = wrad1 / np.pi * fs/2     

    plt.plot( ww1, abs_value1, label = f'alpha_{index}= {this_alpha}')

plt.title( 'Magnitude response' )
plt.xlabel( 'Normalized angular frequency []' )
plt.ylabel( 'Magnitude [dB]' )
plt.ylim( bottom = -50, top = 10 )
plt.grid( visible = 'True', which = 'both', axis = 'both' )
plt.legend()
plt.show()

# ---------------------------------------------------------------------------------------- #
```

###
##### Como se puede observar, la variación de $\alpha$ <u> no produce una variación en el tipo de respuesta del filtro </u> sino que <u> permite ajustar la selectividad del mismo </u>, es decir que, en la medida que $\alpha$ aumenta, mayor será el ancho de las bandas de paso: de forma similar a lo que ocurre con $Q$ en el dominio analógico $S$ de Laplace.
###
##### Por otro lado, podemos observar que <u> se presentan ceros estructurales </u> propios de un filtro notch: <u> la cantidad de ceros está definida por la cantidad de coeficientes, es decir por $M$, y su ubicación en el ancho de banda digital viene dada por su paridad.</u>
###
##### Esto ocurre porque en la medida que aumenta $M$ y por lo tanto la cantidad de coeficientes del filtro, la respuesta se comprime en el ancho de banda digital dado por la frecuencia de Nyquist, incorporando ceros topológicos equiespaciados.
###
##### Para el caso de $M=4$, se observan <u> dos ceros topológicos </u> que se encuentran por caso en $f_1=\frac{1}{4}*\frac{fs}{2}$ y $f_2=\frac{3}{4}*\frac{fs}{2}$, es decir que $f_2 = 3*f_1$: de esta condición nos vamos a valer para anular las frecuencias en cuestión.
###
##### Podemos observar que para anular la señal senoidal interferente de 125Hz, que se corresponderá con $f_1$, y su tercera armónica de 375Hz, que se corresponderá a $f_2$, debemos muestran a la siguiente frecuencia $fs$:
###

\begin{equation*}
f_1= \frac{1}{4}*\frac{fs}{2}= 125Hz 
\end{equation*}

###

\begin{equation*}
fs = 8 * f_1= 8 * 125 Hz = 1000 Hz  
\end{equation*}

###
##### Obtenemos el mismo resultado en caso de hacer lo propio con $f_2$.
###

```python
# ---------------------------------------------------------------------------------------- #

#-- Frecuencia de muestreo --
fs = 1000
f1 = 0.125 * fs             # f1 = 1/4 * fs/2
f2 = 0.375 * fs             # f1 = 3/4 * fs/2

for index, this_alpha in enumerate(alpha):
    
    num = 0.5 * np.array( [ 1+this_alpha, 0, 0, 0, 1+this_alpha ] )
    den = np.array( [ 1, 0, 0, 0, this_alpha ] )

    wrad1, hh1 = sig.freqz( num, den, worN=1000 )    
    abs_value1 = 20 * np.log10( abs( hh1 ) )
    ww1 = wrad1 / np.pi * fs/2     

    plt.plot( ww1, abs_value1, label = f'alpha_{index}= {this_alpha}')


plt.title( 'Magnitude response' )
plt.xlabel( 'Normalized angular frequency []' )
plt.ylabel( 'Magnitude [dB]' )
    
plt.ylim( bottom = -50, top = 10 )    

plt.axvline( x = f1, color = 'red', linestyle = '--', label = f'f1 = {f1}' )
plt.axvline( x = f2, color = 'red', linestyle = '--', label = f'f1 = {f2}' )

plt.text( f1, -55, f'f1 = {f1}', ha = 'center', va = 'center' )
plt.text( f2, -55, f'f2 = {f2}', ha = 'center', va = 'top' )

plt.grid( visible = 'True', which = 'both', axis = 'both' )
plt.legend()
plt.show()

# ---------------------------------------------------------------------------------------- #
```
